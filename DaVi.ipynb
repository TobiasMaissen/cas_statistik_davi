# Weltbevölkerung über die Zeit

```python
# Import der Daten
from owid.catalog import charts
import plotly.express as px
import pandas as pd
import plotly.graph_objects as go

population_df = charts.get_data("population")
```

```python
# Kurzer Einblick in den DF
population_df.head()
```

```python
# Welcher Länder gibt es den?
population_df["entities"].unique()
```

```python
# Wie viele Rows sind dort drin?
population_df
```

```python
# Brauche nur World um die Grafik zu erstellen (und check wie viele Records drin sind)
weltbevoelkerung = population_df[population_df['entities'] == 'World'].copy()
weltbevoelkerung
```

```python
# Umrechnung in Milliarden
weltbevoelkerung['population_billions'] = weltbevoelkerung['population'] / 1_000_000_000

# Erstellen des Plotly-Diagramms mit dunklem Theme
fig = px.line(weltbevoelkerung, 
              x='years', 
              y='population_billions',
              title='Weltbevölkerung über die Zeit',
              template='plotly_dark',
              color_discrete_sequence=px.colors.sequential.Sunsetdark)

# Anpassen der Achsenbeschriftungen und Layout
fig.update_layout(
    xaxis_title="Jahr",
    yaxis_title="Bevölkerung (Milliarden)",
    xaxis=dict(
        range=[-10000, 2023],
        tickvals=[-10000, -4000, 0, 2023],  # Spezifische Jahre auf der x-Achse
        ticktext=['-10000', '-4000', '0', '2023']  # Labels für diese Jahre
    )
)

# Füge Annotationen für die spezifischen Jahre hinzu
for year in [-10000, -4000, 0, 2023]:
    year_data = weltbevoelkerung[weltbevoelkerung['years'] == year]
    if not year_data.empty:
        pop_value = year_data.iloc[0]['population_billions']
        fig.add_annotation(
            x=year,
            y=pop_value,
            text=f"{pop_value:.3f}Mrd",
            showarrow=True,
            arrowhead=1,
            yshift=-2,
            font=dict(color='white')
        )

# Anzeigen des Diagramms
fig.show()
```

```python
import plotly.graph_objects as go

# Daten vorbereiten wie zuvor
historical_data = population_projections_world[population_projections_world['years'] <= 2024].copy()
historical_data['population_billions'] = historical_data['population__sex_all__age_all__variant_estimates'] / 1_000_000_000

forecast_data = population_projections_world[population_projections_world['years'] >= 2024].copy()
forecast_data['population_billions'] = forecast_data['population__sex_all__age_all__variant_medium'] / 1_000_000_000

# Grafik erstellen
fig = go.Figure()

# Historische Daten (kräftiges Grün)
fig.add_trace(go.Scatter(
    x=historical_data['years'],
    y=historical_data['population_billions'],
    name='Historische Daten',
    line=dict(color='#2ecc71', width=3)
))

# Prognosedaten (blasses, gestricheltes Grün)
fig.add_trace(go.Scatter(
    x=forecast_data['years'],
    y=forecast_data['population_billions'],
    name='Prognose',
    line=dict(color='rgba(46, 204, 113, 0.3)', width=2, dash='dash')
))

# Layout anpassen
fig.update_layout(
    title={
        'text': 'Weltbevölkerung: Historie und Prognose',
        'y': 0.95,
        'x': 0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': {'size': 24}
    },
    xaxis_title="Jahr",
    yaxis_title="Bevölkerung (Milliarden)",
    template='plotly_dark',
    showlegend=True,
    legend=dict(
        yanchor="top",
        y=0.99,
        xanchor="left",
        x=0.01
    ),
    height=800,
    width=1200,
    xaxis=dict(
        range=[1950, 2100],     # Bereich der Jahre
        dtick=25,               # Abstand zwischen Tickmarks
        ticks='outside',        # Tickmarks nach außen
        ticklen=10,            # Länge der Tickmarks
        tickfont=dict(size=12), # Schriftgröße der Tickmarks
        tickangle=0,           # Keine Rotation der Jahre
        title_standoff=25,     # Mehr Abstand zwischen Achsentitel und Ticks
        side='bottom'          # Position der Achse unten
    ),
    yaxis=dict(
        range=[2, 11],
        dtick=1,
        gridwidth=0.5,
    ),
    plot_bgcolor='rgba(0,0,0,0.95)',
    paper_bgcolor='rgba(0,0,0,0.95)',
)

# Spezifische Jahre für Annotationen
years_to_annotate = [1950, 2000, 2023]
y_shifts = [20, -30, 30]  # Verschiedene y-Verschiebungen für bessere Lesbarkeit
x_shifts = [50, 50, 50]   # Horizontale Verschiebung für rechtsbündige Ausrichtung

for year, y_shift, x_shift in zip(years_to_annotate, y_shifts, x_shifts):
    data_point = historical_data[historical_data['years'] == year]
    if not data_point.empty:
        fig.add_annotation(
            x=year,
            y=data_point['population_billions'].iloc[0],
            text=f"{year}: {data_point['population_billions'].iloc[0]:.2f} Mrd.",
            showarrow=False,  # Pfeil anzeigen
            font=dict(size=12, color='white'),  # Schriftgröße und Farbe
            bgcolor='rgba(0,0,0,0.7)',  # Hintergrundfarbe der Box
            bordercolor='#2ecc71',  # Farbe des Box-Rands
            borderwidth=1,  # Breite der Box-Rand
            yshift=y_shift,  # Vertikale Verschiebung
            xshift=x_shift,  # Horizontale Verschiebung
            align='left',    # Text-Ausrichtung in der Box
            xanchor='left'   # Ausrichtung der Box relativ zum Datenpunkt
        )

# Maximum-Punkt der Prognose - zentriert
max_point = forecast_data.loc[forecast_data['population_billions'].idxmax()]
fig.add_annotation(
    x=max_point['years'],
    y=max_point['population_billions'],
    text=f"Maximum ({int(max_point['years'])}): {max_point['population_billions']:.2f} Mrd.",
    showarrow=True,  # Pfeil anzeigen
    arrowhead=1,  # Form des Pfeils
    font=dict(size=12, color='white'),  # Schriftgröße und Farbe
    bgcolor='rgba(0,0,0,0.7)',  # Hintergrundfarbe der Box
    bordercolor='rgba(46, 204, 113, 0.3)',
    borderwidth=1,
    yshift=5,        # Vertikale Verschiebung nach oben
    xshift=0,         # Keine horizontale Verschiebung
    align='center',   # Text-Ausrichtung in der Box zentriert
    xanchor='center'  # Box zentriert über dem Datenpunkt
)

# Anzeigen der Grafik
fig.show()
```

```python
import plotly.graph_objects as go

# --- Start of streamlined plotting code ---

# Define known column names
hist_pop_col = 'population__sex_all__age_all__variant_estimates'
forecast_pop_col = 'population__sex_all__age_all__variant_medium'

# Prepare data directly
historical_data = population_projections_china[population_projections_china['years'] <= 2024].copy()
historical_data['population_billions'] = historical_data[hist_pop_col] / 1_000_000_000
historical_data.dropna(subset=['population_billions'], inplace=True)

forecast_data = population_projections_china[population_projections_china['years'] >= 2024].copy()
forecast_data['population_billions'] = forecast_data[forecast_pop_col] / 1_000_000_000
forecast_data.dropna(subset=['population_billions'], inplace=True)

# Create figure
fig = go.Figure()

# Add historical data trace (if data exists)
if not historical_data.empty:
    fig.add_trace(go.Scatter(
        x=historical_data['years'],
        y=historical_data['population_billions'],
        name='Historische Daten (China)',
        line=dict(color='#2ecc71', width=3)
    ))

# Add forecast data trace (if data exists)
if not forecast_data.empty:
    fig.add_trace(go.Scatter(
        x=forecast_data['years'],
        y=forecast_data['population_billions'],
        name='Prognose (China)',
        line=dict(color='rgba(46, 204, 113, 0.3)', width=2, dash='dash')
    ))

# Configure layout
fig.update_layout(
    title={
        'text': 'Bevölkerung China: Historie und Prognose',
        'y': 0.95, 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top',
        'font': {'size': 24}
    },
    xaxis_title="Jahr",
    yaxis_title="Bevölkerung (Milliarden)",
    template='plotly_dark',
    showlegend=True,
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
    height=800,
    width=1200,
    xaxis=dict(
        range=[1950, 2100], dtick=25, ticks='outside', ticklen=10,
        tickfont=dict(size=12), tickangle=0, title_standoff=25, side='bottom'
    ),
    yaxis=dict(range=[0, 1.6], dtick=0.2, gridwidth=0.5),
    plot_bgcolor='rgba(0,0,0,0.95)',
    paper_bgcolor='rgba(0,0,0,0.95)',
)

# Add historical annotations
if not historical_data.empty:
    years_to_annotate_hist = [1950, 2000, 2023] # Keep 2023 here as it's historical peak
    y_shifts_hist = [20, -30, 30]
    x_shifts_hist = [50, 50, 50] # Adjusted x_shift for 2023 to avoid overlap if needed
    for year, y_shift, x_shift in zip(years_to_annotate_hist, y_shifts_hist, x_shifts_hist):
        data_point = historical_data[historical_data['years'] == year]
        if not data_point.empty and pd.notna(data_point['population_billions'].iloc[0]):
            fig.add_annotation(
                x=year, y=data_point['population_billions'].iloc[0],
                text=f"{year}: {data_point['population_billions'].iloc[0]:.2f} Mrd.",
                showarrow=False, font=dict(size=12, color='white'),
                bgcolor='rgba(0,0,0,0.7)', bordercolor='#2ecc71', borderwidth=1,
                yshift=y_shift, xshift=x_shift, align='left', xanchor='left'
            )

# Add forecast annotation for 2100
if not forecast_data.empty:
    year_2100_data = forecast_data[forecast_data['years'] == 2100]
    if not year_2100_data.empty and pd.notna(year_2100_data['population_billions'].iloc[0]):
        pop_2100 = year_2100_data['population_billions'].iloc[0]
        fig.add_annotation(
            x=2100, y=pop_2100,
            text=f"2100: {pop_2100:.2f} Mrd.",
            showarrow=False, font=dict(size=12, color='white'),
            bgcolor='rgba(0,0,0,0.7)', bordercolor='rgba(46, 204, 113, 0.3)', borderwidth=1, # Use forecast color
            yshift=15,      # Position adjustment
            xshift=-50,     # Position adjustment (negative for left alignment)
            align='right',  # Text alignment
            xanchor='right' # Anchor point
        )

# Show plot
fig.show()

# --- End of streamlined plotting code ---
```

## China Vs. USA

```python
import pandas as pd
import numpy as np

# --- Start of Code ---

# Define target entities and years
target_entities = ['China', 'United States'] # Nur China und USA
start_year = 2023
end_year = 2050

# Filter for relevant entities and years
df_filtered = working_age_df[
    (working_age_df['entities'].isin(target_entities)) &
    (working_age_df['years'] >= start_year) &
    (working_age_df['years'] <= end_year)
].copy()

# Define column names
est_col = 'population__sex_all__age_15_64__variant_estimates'
med_col = 'population__sex_all__age_15_64__variant_medium'

# --- Data Preparation ---
# Create a unified population column (using estimates for 2023, medium for forecast)
# Convert to millions directly
df_filtered['working_pop_millions'] = np.where(
    df_filtered['years'] == start_year,
    df_filtered[est_col] / 1_000_000, # Use estimates for start year
    df_filtered[med_col] / 1_000_000  # Use medium forecast for subsequent years
)

# Sort for accurate cumulative calculation
df_filtered.sort_values(by=['entities', 'years'], inplace=True)

# Calculate cumulative change since start_year for each country
# Group by country, calculate diff, then cumulative sum
df_filtered['yearly_change'] = df_filtered.groupby('entities')['working_pop_millions'].diff()
df_filtered['cumulative_change_millions'] = df_filtered.groupby('entities')['yearly_change'].cumsum().fillna(0)

# Select and pivot the table
comparison_df = df_filtered.pivot_table(
    index='years',
    columns='entities',
    values=['working_pop_millions', 'cumulative_change_millions']
)

# Flatten the multi-level column index for better readability
comparison_df.columns = [f"{col[1]}_{col[0].replace('_millions', '').replace('working_pop','Pop').replace('cumulative_change','Cumul_Change')}" for col in comparison_df.columns]

# Reorder columns for clarity (optional)
comparison_df = comparison_df[[
    'China_Pop', 'China_Cumul_Change', 
    'United States_Pop', 'United States_Cumul_Change'
]]

# Round the values for display
comparison_df = comparison_df.round(2)

# Display the resulting DataFrame
print(f"Working-Age Population Comparison (Millions): {start_year}-{end_year}")
comparison_df

# --- End of Code ---
```

Im 2050 hat ``China eine kumulative Veränderung von -237 Millionen an Working-Age Personen``, was in etwa dem entspricht, was die komplette USA zum Stand 2023 hat... Crazy

```python
import plotly.graph_objects as go
import pandas as pd
import numpy as np

# --- Annahmen ---
# 1. DataFrame 'working_age_df' existiert.
# 2. Benötigte Spaltennamen:
hist_work_pop_col = 'population__sex_all__age_15_64__variant_estimates'
forecast_work_pop_col = 'population__sex_all__age_15_64__variant_medium'

# --- Start des Codes für die Grafik ---

# Filtere das working_age_df nach China
try:
    working_age_china = working_age_df[working_age_df['entities'] == 'China'].copy()
    if working_age_china.empty:
        print("Warnung: Keine Daten für 'China' im working_age_df gefunden.")
        fig = go.Figure()
        fig.update_layout(title="Fehler: Keine Daten für China gefunden", template='plotly_dark')
        fig.show()
        # exit()
except NameError:
    print("Fehler: Das DataFrame 'working_age_df' wurde nicht gefunden. Bitte lade die Daten zuerst.")
    fig = go.Figure()
    fig.update_layout(title="Fehler: 'working_age_df' nicht gefunden", template='plotly_dark')
    fig.show()
    # exit()
except Exception as e:
    print(f"Ein unerwarteter Fehler beim Filtern des DataFrames ist aufgetreten: {e}")
    fig = go.Figure()
    fig.update_layout(title="Fehler: Beim Filtern der Daten", template='plotly_dark')
    fig.show()
    # exit()

# Nur fortfahren, wenn working_age_china Daten enthält
if not working_age_china.empty:

    # Überprüfe Spalten im gefilterten DataFrame working_age_china
    if hist_work_pop_col not in working_age_china.columns or forecast_work_pop_col not in working_age_china.columns:
        print(f"*** Fehler: Benötigte Spalten nicht im working_age_china DataFrame gefunden:")
        # ... (Fehlermeldung wie zuvor) ...
        fig = go.Figure()
        fig.update_layout(title="Fehler: Benötigte Spalten fehlen", template='plotly_dark')
    else:
        # Daten vorbereiten (aus working_age_china)
        work_hist_data = working_age_china[working_age_china['years'] <= 2024].copy()
        if not work_hist_data.empty and hist_work_pop_col in work_hist_data.columns:
            work_hist_data['work_pop_billions'] = work_hist_data[hist_work_pop_col] / 1_000_000_000
            work_hist_data.dropna(subset=['work_pop_billions'], inplace=True)
        else:
            work_hist_data = pd.DataFrame() # Leeres DataFrame

        work_forecast_data = working_age_china[working_age_china['years'] >= 2024].copy()
        if not work_forecast_data.empty and forecast_work_pop_col in work_forecast_data.columns:
           work_forecast_data['work_pop_billions'] = work_forecast_data[forecast_work_pop_col] / 1_000_000_000
           work_forecast_data.dropna(subset=['work_pop_billions'], inplace=True)
        else:
           work_forecast_data = pd.DataFrame() # Leeres DataFrame

        # --- Wert für 2023 als Referenz holen ---
        value_2023 = 0
        year_2023 = 2023
        # Suche in historischen Daten, falls vorhanden, sonst in Prognosedaten (für den Fall, dass 2023 als Prognose zählt)
        data_2023 = pd.DataFrame()
        if not work_hist_data.empty and 'work_pop_billions' in work_hist_data.columns:
             data_2023 = work_hist_data[work_hist_data['years'] == year_2023]
        
        # Falls nicht in historischen Daten gefunden oder 'work_pop_billions' fehlt, suche in Prognosedaten
        if data_2023.empty and not work_forecast_data.empty and 'work_pop_billions' in work_forecast_data.columns:
             data_2023 = work_forecast_data[work_forecast_data['years'] == year_2023]

        if not data_2023.empty:
            value_2023 = data_2023['work_pop_billions'].iloc[0]
            print(f"Referenzwert für {year_2023}: {value_2023:.3f} Mrd.")
        else:
            print(f"Keine Daten für das Jahr {year_2023} gefunden. Vergleich nicht möglich.")
            # Setze value_2023 auf NaN oder 0, um Berechnungsfehler unten zu vermeiden
            value_2023 = np.nan # Oder 0, je nachdem wie der Rest behandelt werden soll


        # Wert für 2050 holen
        value_2050 = 0
        year_2050 = 2050
        if not work_forecast_data.empty and 'work_pop_billions' in work_forecast_data.columns:
            data_2050 = work_forecast_data[work_forecast_data['years'] == year_2050]
            if not data_2050.empty:
                value_2050 = data_2050['work_pop_billions'].iloc[0]
                print(f"Wert für {year_2050}: {value_2050:.3f} Mrd.")
            else:
                 print(f"Keine Daten für das Jahr {year_2050} in der Prognose gefunden.")
                 value_2050 = np.nan # Setze auf NaN, falls nicht gefunden
        else:
            print("Keine Prognosedaten zur Bestimmung des Werts für 2050 vorhanden.")
            value_2050 = np.nan

        # Prozentuale Veränderung berechnen (2023 -> 2050)
        percentage_change_23_50 = 0
        # Stelle sicher, dass beide Werte gültige Zahlen sind
        if pd.notna(value_2023) and pd.notna(value_2050) and value_2023 != 0:
            percentage_change_23_50 = round((value_2050 - value_2023) / value_2023 * 100)
            print(f"Prozentuale Veränderung von {year_2023} bis {year_2050}: {percentage_change_23_50:+}%")
        elif value_2023 == 0:
            print(f"Prozentuale Veränderung kann nicht berechnet werden (Wert für {year_2023} ist 0).")
        else:
             print(f"Prozentuale Veränderung kann nicht berechnet werden (Wert für {year_2023} oder {year_2050} fehlt).")


        # --- Grafik erstellen ---
        fig = go.Figure()

        # Historische Daten hinzufügen
        if not work_hist_data.empty and 'work_pop_billions' in work_hist_data.columns:
            fig.add_trace(go.Scatter(
                x=work_hist_data['years'],
                y=work_hist_data['work_pop_billions'],
                mode='lines', name='Historisch (15-64 J.)',
                line=dict(color='#2ecc71', width=3)
            ))

        # Prognosedaten hinzufügen
        if not work_forecast_data.empty and 'work_pop_billions' in work_forecast_data.columns:
            fig.add_trace(go.Scatter(
                x=work_forecast_data['years'],
                y=work_forecast_data['work_pop_billions'],
                mode='lines', name='Prognose (15-64 J.)',
                line=dict(color='rgba(46, 204, 113, 0.5)', width=2, dash='dash')
            ))

        # --- Horizontale Linie ab 2023 ---
        # Nur zeichnen, wenn beide Werte gültig sind
        if pd.notna(value_2023) and pd.notna(value_2050):
             fig.add_trace(go.Scatter(
                 x=[year_2023, year_2050], # Start bei 2023
                 y=[value_2023, value_2023], # Auf Höhe von 2023
                 mode='lines',
                 name=f'Niveau {year_2023}',
                 line=dict(color='rgba(255, 0, 0, 0.7)', width=2, dash='dot'),
                 showlegend=False
             ))

        # --- Vertikale Linie zwischen 2023 und 2050 Wert ---
        # Nur zeichnen, wenn beide Werte gültig sind
        if pd.notna(value_2023) and pd.notna(value_2050):
            fig.add_shape(type="line",
                x0=year_2050, y0=value_2050, x1=year_2050, y1=value_2023, # Spanne von 2050 Wert bis 2023 Wert
                line=dict(color="rgba(255, 0, 0, 0.7)",width=2)
            )
            # Kleine Ticks oben/unten an der vertikalen Linie
            fig.add_shape(type="line",
                x0=year_2050-0.5, y0=value_2023, x1=year_2050+0.5, y1=value_2023, # Tick auf Höhe 2023
                line=dict(color="rgba(255, 0, 0, 0.7)", width=2)
            )
            fig.add_shape(type="line",
                x0=year_2050-0.5, y0=value_2050, x1=year_2050+0.5, y1=value_2050, # Tick auf Höhe 2050
                line=dict(color="rgba(255, 0, 0, 0.7)", width=2)
            )

        # --- Layout und Styling ---
        fig.update_layout(
             title={
                'text': 'Arbeitsfähige Bevölkerung China (15-64 J.): Historie und Prognose',
                'y': 0.95, 'x': 0.5, 'xanchor': 'center', 'yanchor': 'top',
                'font': {'size': 20}
            },
            xaxis_title="Jahr",
            yaxis_title="Bevölkerung (Milliarden)",
            template='plotly_dark',
            showlegend=True,
            legend=dict(
                yanchor="bottom", y=0.01, xanchor="left", x=0.01,
                bgcolor='rgba(0,0,0,0.5)'
                ),
            height=700, width=1200,
            xaxis=dict(
                range=[1980, 2060], dtick=20, ticks='outside', ticklen=10,
                tickfont=dict(size=12), tickangle=0, title_standoff=25, side='bottom',
                gridcolor='rgba(255, 255, 255, 0.2)', linecolor='rgba(255, 255, 255, 0.5)',
                zerolinecolor='rgba(255, 255, 255, 0.3)'
            ),
            yaxis=dict(
                range=[0, 1.1], dtick=0.2, gridwidth=0.5,
                gridcolor='rgba(255, 255, 255, 0.2)', linecolor='rgba(255, 255, 255, 0.5)',
                zerolinecolor='rgba(255, 255, 255, 0.3)'
            ),
            plot_bgcolor='rgba(0,0,0,0.95)', paper_bgcolor='rgba(0,0,0,0.95)',
            font=dict(color='white')
        )

        # --- Annotation für prozentuale Veränderung ---
        # Nur hinzufügen, wenn Berechnung erfolgreich war und Werte gültig sind
        if percentage_change_23_50 != 0 and pd.notna(value_2023) and pd.notna(value_2050):
            fig.add_annotation(
                x=year_2050,
                y=(value_2023 + value_2050) / 2, # Position in der Mitte der vertikalen Linie (zwischen 2023 und 2050 Wert)
                text=f"<b>{percentage_change_23_50:+}%</b>", # Mit Vorzeichen (+/-)
                showarrow=False,
                font=dict(color="red", size=14), # Rote Schrift
                bgcolor='rgba(0,0,0,0.6)', # Dunkler Hintergrund für Text
                bordercolor='red',
                borderwidth=1,
                xshift=35 # Text etwas nach rechts verschieben
            )
        elif pd.notna(value_2023) and pd.notna(value_2050): # Fall abdecken, wo Änderung 0 ist
             print("Info: Prozentuale Veränderung ist 0%. Annotation wird nicht hinzugefügt.")

        # Plot anzeigen (nur wenn Daten vorhanden waren)
        fig.show()

# --- Ende des Codes ---
```

```python
# Berechnen des Rückgangs der Geburtenraten für jedes Land
def calculate_fertility_decline(df):
    # Gruppieren nach Ländern
    countries = df['entities'].unique()
    decline_data = []
    
    for country in countries:
        country_data = df[df['entities'] == country].sort_values('years')
        
        # Prüfen, ob genügend Daten vorhanden sind
        if len(country_data) > 10:  # Mindestens 10 Datenpunkte für eine sinnvolle Analyse
            # Frühester und spätester verfügbarer Wert
            earliest = country_data.iloc[0]
            latest = country_data.iloc[-1]
            
            # Berechnen des absoluten und prozentualen Rückgangs
            absolute_decline = earliest['children_per_woman_un'] - latest['children_per_woman_un']
            percent_decline = (absolute_decline / earliest['children_per_woman_un']) * 100
            
            # Zeitraum
            year_span = latest['years'] - earliest['years']
            
            # Durchschnittlicher jährlicher Rückgang
            annual_decline = absolute_decline / year_span if year_span > 0 else 0
            
            decline_data.append({
                'country': country,
                'start_year': earliest['years'],
                'end_year': latest['years'],
                'start_rate': earliest['children_per_woman_un'],
                'end_rate': latest['children_per_woman_un'],
                'absolute_decline': absolute_decline,
                'percent_decline': percent_decline,
                'year_span': year_span,
                'annual_decline': annual_decline
            })
    
    return pd.DataFrame(decline_data)

# Berechnen des Rückgangs
decline_df = calculate_fertility_decline(fertility_rate)
```

```python
# Sortieren nach absolutem Rückgang (absteigend)
top_absolute_decline = decline_df.sort_values('absolute_decline', ascending=False).head(5)
print("Top 5 Länder mit dem größten absoluten Rückgang der Geburtenrate:")
top_absolute_decline[['country', 'start_year', 'end_year', 'start_rate', 'end_rate', 'absolute_decline']]
```

```python
# Sortieren nach prozentualem Rückgang (absteigend)
top_percent_decline = decline_df.sort_values('percent_decline', ascending=False).head(5)
print("\nTop 5 Länder mit dem größten prozentualen Rückgang der Geburtenrate:")
top_percent_decline[['country', 'start_year', 'end_year', 'start_rate', 'end_rate', 'percent_decline']]
```

**Fazit**

- Wenn man alle Faktoren (Altersverteilung, Sterblichkeit, Migration, altersabhängige Fertilität, etc.) berücksichtigen will, kommt man um ein `Cohort-Component-Modell (bzw. Leslie-Matrix-Modell)` nicht herum.
- Eine **einfache Gleichung existiert nicht, weil die Realität zu komplex ist**. Stattdessen iteriert man das Modell über mehrere Zeitschritte, oft in 1- oder 5-Jahres-Intervallen.
- Genau solche Modelle werden von Institutionen wie der UN, der Weltbank oder nationalen Statistikämtern genutzt, um Bevölkerungsprognosen bis 2100 oder darüber hinaus zu erstellen.

```python
import pandas as pd
import numpy as np # Wird für einige Operationen benötigt, z.B. bin midpoints
import plotly.graph_objects as go

# Annahme: diamonds_df ist bereits aus der vorherigen Zelle geladen
# diamonds_df = sns.load_dataset('diamonds') # Falls nicht, diese Zeile einkommentieren

# 1. Variablen auswählen
x_var = 'price'
y_var = 'carat'
z_var = 'z'

# Kopie erstellen, um Original-DataFrame nicht zu verändern
df_for_surface = diamonds_df.copy()

# 2. Bins für x und y erstellen
num_bins = 25 # Anzahl der Bins für die x- und y-Achse (kann angepasst werden)
df_for_surface['x_bins'] = pd.cut(df_for_surface[x_var], bins=num_bins)
df_for_surface['y_bins'] = pd.cut(df_for_surface[y_var], bins=num_bins)

# 3. Nach Bins gruppieren und den Mittelwert von z_var berechnen
# observed=False stellt sicher, dass alle Bins im Ergebnis erscheinen, auch wenn sie leer sind
surface_data_grouped = df_for_surface.groupby(['y_bins', 'x_bins'], observed=False)[z_var].mean()
surface_data_unstacked = surface_data_grouped.unstack()

# 4. x und y Koordinaten für die Achsen (Mittelpunkte der Bins)
# Sicherstellen, dass die Kategorien auch dann vorhanden sind, wenn surface_data_unstacked leer sein sollte oder Indizes fehlen
x_bin_categories = df_for_surface['x_bins'].cat.categories
y_bin_categories = df_for_surface['y_bins'].cat.categories

x_coords = [interval.mid for interval in x_bin_categories]
y_coords = [interval.mid for interval in y_bin_categories]

# Die Z-Werte müssen den Dimensionen von y_coords (Zeilen) und x_coords (Spalten) entsprechen.
# Wir müssen die surface_data_unstacked möglicherweise neu indizieren, um sicherzustellen, dass sie die volle Gitterstruktur hat,
# falls einige Kombinationen von Bins keine Daten hatten (was zu NaNs führt, die Plotly handhaben kann).
z_values = surface_data_unstacked.reindex(index=y_bin_categories, columns=x_bin_categories).values

# 5. Das 3D-Oberflächendiagramm erstellen
fig = go.Figure(data=[go.Surface(
    z=z_values,
    x=x_coords,
    y=y_coords,
    colorscale='Viridis', # Farbskala kann angepasst werden
    colorbar_title=f'Mean {z_var}'
)])

fig.update_layout(
    title=f'3D Surface Plot: Mean {z_var} vs. {x_var} and {y_var}',
    scene=dict(
        xaxis_title=x_var,
        yaxis_title=y_var,
        zaxis_title=f'Mean {z_var}',
        # Aspect ratio anpassen, falls nötig
        # aspectmode='cube' # oder 'data', 'manual'
    ),
    margin=dict(l=0, r=0, b=0, t=40) # Ränder anpassen
)

fig.show()
```

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from scipy.interpolate import griddata

# Annahme: diamonds_df ist bereits aus der vorherigen Zelle geladen
# import seaborn as sns
# diamonds_df = sns.load_dataset('diamonds') # Falls nicht, diese Zeile einkommentieren

# 1. Variablen auswählen (NEUE ZUWEISUNG)
x_var = 'price' # Preis auf der x-Achse
y_var = 'carat' # Karat auf der y-Achse
z_var = 'z'     # 'z'-Dimension des Diamanten als Oberflächenhöhe

# Kopie erstellen, um Original-DataFrame nicht zu verändern
df_for_surface = diamonds_df.copy()

# Filter für gültige 'z' Werte (z > 0)
df_for_surface = df_for_surface[df_for_surface['z'] > 0]

# 2. Bins für x ('price') und y ('carat') erstellen
num_bins = 25
df_for_surface['x_bins'] = pd.cut(df_for_surface[x_var], bins=num_bins, labels=False, include_lowest=True)
df_for_surface['y_bins'] = pd.cut(df_for_surface[y_var], bins=num_bins, labels=False, include_lowest=True)

x_bin_edges = pd.cut(df_for_surface[x_var], bins=num_bins, include_lowest=True).cat.categories
y_bin_edges = pd.cut(df_for_surface[y_var], bins=num_bins, include_lowest=True).cat.categories

x_coords_orig = np.array([interval.mid for interval in x_bin_edges])
y_coords_orig = np.array([interval.mid for interval in y_bin_edges])

# 3. Nach Bins gruppieren und den Mittelwert von z_var ('z'-Dimension) berechnen
surface_data_grouped = df_for_surface.groupby(['y_bins', 'x_bins'], observed=False)[z_var].mean()
z_values_orig = surface_data_grouped.unstack().reindex(
    index=range(num_bins), columns=range(num_bins)
).values

# 4. Interpolation für eine glattere Oberfläche
X_orig, Y_orig = np.meshgrid(x_coords_orig, y_coords_orig)
points = np.array([X_orig.ravel(), Y_orig.ravel()]).T
values = z_values_orig.ravel()

valid_points = points[~np.isnan(values)]
valid_values = values[~np.isnan(values)]

if len(valid_points) < 3:
    print(f"Nicht genügend gültige Datenpunkte für die Interpolation (basierend auf Mittelwerten von '{z_var}'). Der Plot könnte leer sein oder Fehler aufweisen.")
    x_coords_interp = np.array([])
    y_coords_interp = np.array([])
    z_values_interp = np.array([[]])
else:
    num_interp_points = 100
    x_coords_interp = np.linspace(x_coords_orig.min(), x_coords_orig.max(), num_interp_points)
    y_coords_interp = np.linspace(y_coords_orig.min(), y_coords_orig.max(), num_interp_points)
    X_interp, Y_interp = np.meshgrid(x_coords_interp, y_coords_interp)
    z_values_interp = griddata(valid_points, valid_values, (X_interp, Y_interp), method='cubic')

# 5. Das 3D-Oberflächendiagramm erstellen mit den interpolierten Daten
fig = go.Figure(data=[go.Surface(
    z=z_values_interp,
    x=x_coords_interp,
    y=y_coords_interp,
    colorscale='Viridis',
    colorbar_title=f'Mean {z_var} (interpolated)'
)])

fig.update_layout(
    title=f'Smoothed 3D Surface Plot: Mean Diamond {z_var}-dimension vs. {x_var} and {y_var}',
    scene=dict(
        xaxis_title=x_var,
        yaxis_title=y_var,
        zaxis_title=f'Mean {z_var}-dimension (mm)',
        camera=dict(
            up=dict(x=0, y=0, z=1),          # Stellt sicher, dass die Z-Achse "oben" ist
            eye=dict(x=-1.8, y=-1.8, z=1.5), # Ändert die Kameraposition
                                             # Negative x,y Werte schauen aus einem anderen Quadranten auf den Ursprung
                                             # Ein z-Wert > 1 schaut tendenziell mehr "von oben"
            center=dict(x=0, y=0, z=0)       # Definiert den Punkt, auf den die Kamera blickt (oft der Ursprung der Szene)
        )
    ),
    margin=dict(l=0, r=0, b=0, t=40)
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# This assumes df_bookings is already loaded from 'bookings.csv'
# and contains columns: 'Room', 'City', 'Year', 'Count'.

# Ensure 'Year' is string and 'Count' is numeric.
df_bookings['Year'] = df_bookings['Year'].astype(str)
df_bookings['Count'] = pd.to_numeric(df_bookings['Count'])

# --- 1. Datenaggregation und -vorbereitung ---
# Remove 'Total' from the cities as requested
df_filtered = df_bookings[df_bookings['City'] != 'Total'].copy()

# Kumulierte Zimmertypen: Summe der Buchungen pro Stadt und Jahr
df_agg = df_filtered.groupby(['City', 'Year'])['Count'].sum().unstack(fill_value=0)

# Ensure both year columns exist
if '2019' not in df_agg.columns:
    df_agg['2019'] = 0
if '2022' not in df_agg.columns:
    df_agg['2022'] = 0

# Sort by 2022 bookings in descending order
df_agg = df_agg.sort_values(by='2022', ascending=False)

df_agg = df_agg.reset_index()

# The city order for the y-axis will now be determined by the sort_values above
# We extract this order directly from the sorted DataFrame
city_order_sorted_by_2022 = df_agg['City'].tolist()

# --- 2. Plot erstellen ---
fig = go.Figure()

# Balken für 2022 (rot)
fig.add_trace(go.Bar(
    y=df_agg['City'],
    x=df_agg['2022'],
    name='2022 Bookings',
    orientation='h',
    marker_color='red'
))

# Markierungsstrich (als Scatter-Marker) für 2019 (blau)
fig.add_trace(go.Scatter(
    y=df_agg['City'],
    x=df_agg['2019'],
    name='2019 Bookings',
    mode='markers',
    marker=dict(
        color='blue',
        symbol='line-ns',
        size=15,
        line=dict(
            color='blue',
            width=3
        )
    )
))

# --- 3. Layout anpassen ---
fig.update_layout(
    title_text='City Bookings: 2022 (Bar) vs 2019 (Marker) - Sorted by 2022',
    title_x=0.5,
    xaxis_title='Number of Bookings (All Room Types Combined)',
    yaxis_title='City',
    legend_title_text='Reference',
    height=max(500, len(city_order_sorted_by_2022) * 35 + 100),
    yaxis=dict(
        # Set the category order for the y-axis to the new sorted list
        categoryorder='array',
        categoryarray=city_order_sorted_by_2022
    ),
    bargap=0.3
)

fig.show()
```
