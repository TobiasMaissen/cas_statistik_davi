## Schritt 1: Datenbereinigung und -aufbereitung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Annahme: df_2020_final ist Ihr DataFrame
df_clean = df_2020_final.copy()

# Liste der Spalten, in denen negative Werte NaN bedeuten könnten
# (Passen Sie diese Liste ggf. an, basierend auf Ihrem Codebuch)
cols_to_clean = [col for col in df_clean.columns if col.startswith('Q')]
cols_to_clean.extend(['X003R']) # Fügen Sie weitere demografische Spalten hinzu, falls nötig

# Ersetzen Sie negative Werte durch NaN
for col in cols_to_clean:
    # Sicherstellen, dass die Spalte numerisch ist, bevor wir vergleichen
    if pd.api.types.is_numeric_dtype(df_clean[col]):
        df_clean.loc[df_clean[col] < 0, col] = np.nan
    else:
        # Versuch, die Spalte in numerisch umzuwandeln, falls sie als Objekt gelesen wurde
        try:
            numeric_col = pd.to_numeric(df_clean[col], errors='coerce')
            df_clean[col] = numeric_col # Originalspalte überschreiben
            df_clean.loc[df_clean[col] < 0, col] = np.nan
            print(f"Spalte {col} wurde in numerisch konvertiert und bereinigt.")
        except Exception as e:
            print(f"Spalte {col} konnte nicht sicher bereinigt werden: {e}")


# Spezifische Bereinigung für Q260 (Geschlecht), falls nötig (Codebuch prüfen!)
if 'Q260' in df_clean.columns and pd.api.types.is_numeric_dtype(df_clean['Q260']):
     df_clean.loc[df_clean['Q260'] < 0, 'Q260'] = np.nan
```

**2. QQ-Plots, t-Tests, Chi-Quadrat-Tests etc.**
Deine Verwirrung ist absolut verständlich, diese statistischen Tests können am Anfang überwältigend sein. Lass uns das kurz einordnen:

- **Was wir gerade machen:** Deskriptive Statistik. Wir beschreiben, wie die Daten aussehen (Häufigkeiten, Mittelwerte, Streuungen, Verteilungen durch Histogramme/Boxplots). Ziel ist es, ein Gefühl für die Daten zu bekommen.

- **Was diese Tests machen:** Inferenzstatistik. Sie werden verwendet, um Hypothesen zu testen oder Rückschlüsse von der Stichprobe auf die Grundgesamtheit zu ziehen.
  - **QQ-Plot:** Ein visuelles Werkzeug, um zu prüfen, ob Daten einer bestimmten theoretischen Verteilung folgen (oft Normalverteilung). Nützlich nachdem man sich die Verteilung z.B. im Histogramm angesehen hat, um die Abweichung von der Normalität genauer zu beurteilen. Könnten wir z.B. für Q262 (Alter) machen, um die Abweichung von der Normalverteilung deutlicher zu sehen.
  - **t-Test:** Vergleicht die Mittelwerte einer (ungefähr) normalverteilten Variable zwischen zwei Gruppen. Beispiel: "Ist der durchschnittliche Q163-Score bei Männern signifikant anders als bei Frauen?" (Voraussetzungen wie Normalverteilung innerhalb der Gruppen und Varianzhomogenität müssen geprüft werden, bei grossen Stichproben ist der Test aber robust gegen Verletzungen der Normalverteilungsannahme).
  - **Chi-Quadrat-Test (χ²):** Prüft, ob ein Zusammenhang zwischen zwei kategorialen Variablen besteht. Beispiel: "Gibt es einen Zusammenhang zwischen der Altersgruppe (X003R) und dem Land (B_COUNTRY_ALPHA)?" oder "Gibt es einen Zusammenhang zwischen Geschlecht (Q260) und einer kategorialen Antwort auf eine Frage?".
  
Wann verwenden? Diese Tests kommen meist nach der deskriptiven Analyse ins Spiel, wenn du spezifische Fragen oder Hypothesen hast, die du testen möchtest.
Beispiel: Nachdem wir sehen, dass Männer und Frauen leicht unterschiedliche Durchschnittswerte bei Q163 haben (deskriptiv), könnten wir einen t-Test verwenden, um zu prüfen, ob dieser Unterschied statistisch signifikant ist (inferenziell).
Beispiel: Nachdem wir sehen, dass die Verteilung der Altersgruppen je nach Land unterschiedlich aussieht (deskriptiv in Kreuztabellen), könnten wir einen Chi-Quadrat-Test verwenden, um zu prüfen, ob dieser Zusammenhang signifikant ist (inferenziell).

```python
# --- Analyse: Altersgruppenverteilung pro Land (Heatmap) ---
print("\n--- Analyse: Altersgruppenverteilung pro Land (Heatmap) ---")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Für np.nan wenn Daten fehlen

# Sicherstellen, dass df_clean und die Spalten existieren
if 'df_clean' not in locals() or 'X003R' not in df_clean.columns or 'B_COUNTRY_ALPHA' not in df_clean.columns:
    print("Fehler: Benötigte Daten (df_clean, X003R, B_COUNTRY_ALPHA) nicht gefunden.")
else:
    # Mapping und Reihenfolge für Altersgruppen-Labels
    age_labels = {1.0: '16-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54', 5.0: '55-64', 6.0: '65+'}
    ordered_age_labels = list(age_labels.values())

    # Daten vorbereiten (wie zuvor)
    df_age_country = df_clean[['B_COUNTRY_ALPHA', 'X003R']].dropna()
    df_age_country = df_age_country[df_age_country['X003R'].isin(age_labels.keys())]
    df_age_country['Altersgruppe_Label'] = df_age_country['X003R'].map(age_labels)

    # Kreuztabelle mit absoluten Zahlen
    crosstab_abs = pd.crosstab(df_age_country['B_COUNTRY_ALPHA'], df_age_country['Altersgruppe_Label'])
    crosstab_abs = crosstab_abs.reindex(columns=ordered_age_labels, fill_value=0)

    # Kreuztabelle in Prozent umrechnen
    # Füge kleine Konstante hinzu, um Division durch Null bei Ländern ohne Antworten zu vermeiden
    row_sums = crosstab_abs.sum(axis=1)
    # Länder ohne Antworten (Summe=0) identifizieren und für diese Zeile NaN setzen
    crosstab_perc = crosstab_abs.apply(lambda x: (x / row_sums[x.name] * 100) if row_sums[x.name] > 0 else np.nan, axis=1)

    # Optional: Länder nach Gesamtanzahl sortieren (oder nach einem anderen Kriterium)
    country_counts_sorted_index = df_clean['B_COUNTRY_ALPHA'].value_counts().index
    crosstab_perc_sorted = crosstab_perc.reindex(country_counts_sorted_index)
    # Entferne Zeilen, die nur NaN enthalten (Länder ohne gültige Altersdaten)
    crosstab_perc_sorted.dropna(how='all', inplace=True)


    # --- Visualisierung: Heatmap ---
    if not crosstab_perc_sorted.empty:
        plt.figure(figsize=(12, 10)) # Grösse anpassen je nach Anzahl Länder
        sns.heatmap(
            crosstab_perc_sorted,
            annot=True,       # Zeige die Prozentwerte in den Zellen
            fmt=".1f",        # Formatierung der Werte (eine Nachkommastelle)
            linewidths=.5,    # Linien zwischen Zellen
            cmap="viridis",   # Farbpalette (andere Optionen: "Blues", "YlGnBu", "coolwarm", etc.)
            cbar_kws={'label': 'Anteil der Befragten (%)'} # Beschriftung der Farbleiste
        )
        plt.title('Heatmap der relativen Altersgruppenverteilung pro Land', fontsize=16)
        plt.xlabel('Altersgruppe', fontsize=12)
        plt.ylabel('Land (ALPHA-Code)', fontsize=12)
        plt.xticks(rotation=45, ha='right') # X-Achsen-Labels lesbarer machen
        plt.yticks(rotation=0) # Y-Achsen-Labels horizontal
        plt.tight_layout()
        plt.show()
    else:
        print("Keine ausreichenden Daten zur Visualisierung der Altersgruppenverteilung pro Land vorhanden.")

    print("-" * 70)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np # Sicherstellen, dass numpy importiert ist

# --- Univariate Analyse - Teil 2: Ausgewählte Q-Fragen ---

print("--- Analyse ausgewählter Q-Fragen ---")

# Hilfsfunktion für die Analyse (jetzt mit optionalem ax Parameter)
# Gibt den (potenziell modifizierten) DataFrame zurück
def analyze_question(df, q_col, q_text, labels=None, scale_type='ordinal', ax=None):
    """
    Analysiert eine einzelne Frage (Spalte) im DataFrame.

    Args:
        df (pd.DataFrame): Der DataFrame.
        q_col (str): Der Spaltenname der Frage.
        q_text (str): Beschreibender Text für die Frage (für Titel).
        labels (dict, optional): Mapping von numerischen Werten zu Textlabels. Defaults to None.
        scale_type (str, optional): Typ der Skala ('ordinal', 'scale_1_10', 'binary'). Beeinflusst zusätzliche Plots. Defaults to 'ordinal'.
        ax (matplotlib.axes.Axes, optional): Achsenobjekt zum Plotten. Wenn None, wird eine neue Figur erstellt. Defaults to None.

    Returns:
        pd.DataFrame: Der potenziell modifizierte DataFrame (z.B. durch Typkonvertierung).
    """
    print(f"\n--- Analyse von {q_col}: {q_text} ---")
    if q_col not in df.columns:
        print(f"FEHLER: Spalte {q_col} nicht im DataFrame gefunden.")
        # Gebe das Original-df zurück, um Fehler im weiteren Verlauf zu vermeiden
        return df

    # Kopie erstellen, um SettingWithCopyWarning zu vermeiden und Original-df nicht zu ändern
    df_copy = df.copy()

    # Sicherstellen, dass die Spalte numerisch ist, falls nicht, versuchen zu konvertieren
    if not pd.api.types.is_numeric_dtype(df_copy[q_col]):
         try:
             # Versuche, in numerischen Typ zu konvertieren, erzwinge Fehler zu NaN
             df_copy[q_col] = pd.to_numeric(df_copy[q_col], errors='coerce')
             print(f"Hinweis: Spalte {q_col} wurde zu numerischem Typ konvertiert.")
         except Exception as e:
              print(f"FEHLER: Konnte Spalte {q_col} nicht in numerischen Typ konvertieren: {e}")
              # Gebe das Original-df zurück
              return df

    # Fehlende Werte zählen
    nan_count = df_copy[q_col].isnull().sum()
    total_count = len(df_copy[q_col])
    print(f"Anzahl fehlender Werte (NaN): {nan_count} (von {total_count} Einträgen)")

    # Value Counts (Häufigkeiten der Antworten)
    print("\nVerteilung der Antworten (Value Counts):")
    value_counts = df_copy[q_col].value_counts(dropna=False).sort_index() # dropna=False zeigt NaN count
    print(value_counts)

    # Deskriptive Statistik (nur sinnvoll für numerische Daten)
    print("\nDeskriptive Statistiken:")
    # Runden für bessere Lesbarkeit
    print(df_copy[q_col].describe().round(2))

    # --- Plotting ---
    # Entscheiden, ob eine neue Figur erstellt werden soll oder das übergebene ax verwendet wird
    create_new_figure = ax is None

    if create_new_figure:
        fig, current_ax = plt.subplots(figsize=(8, 5))
    else:
        current_ax = ax # Verwende das übergebene Axes-Objekt

    # Temporäre Label-Spalte nur erstellen, wenn Labels vorhanden sind
    temp_label_col = f'{q_col}_label_temp'
    plot_data = df_copy.dropna(subset=[q_col]) # Nur gültige Werte für Plot verwenden

    if labels:
        # Erstelle die temporäre Spalte nur im Plot-Daten-Frame
        plot_data = plot_data.copy() # Vermeide Warnung
        plot_data[temp_label_col] = plot_data[q_col].map(labels)

        # Sortiere die Labels basierend auf den ursprünglichen numerischen Werten
        valid_keys = sorted([k for k in labels.keys() if pd.notna(k)])
        ordered_labels = [labels[key] for key in valid_keys]

        sns.countplot(data=plot_data, x=temp_label_col, order=ordered_labels, palette='viridis', ax=current_ax)
        current_ax.set_xlabel("Antwort")
        # Ticks rotieren nur bei Textlabels
        current_ax.tick_params(axis='x', rotation=45, labelsize=9 if not create_new_figure else 10)

    else:
        # Direkt mit numerischen Werten plotten, wenn keine Labels vorhanden
        sns.countplot(data=plot_data, x=q_col, palette='viridis', ax=current_ax)
        current_ax.set_xlabel(f"Antwort ({q_col})")
        # Keine Rotation für numerische Ticks

    current_ax.set_ylabel("Anzahl")
    # Kürzeren Titel für Subplots, längeren für Einzelplots
    plot_title = f"{q_col}: {q_text}" if not create_new_figure else f"Verteilung der Antworten für {q_col}: {q_text}"
    current_ax.set_title(plot_title, fontsize=10 if not create_new_figure else 12)


    # Zusätzliche Plots (Histogramm, Boxplot) nur, wenn eine *neue* Figur erstellt wurde
    if create_new_figure:
        plt.tight_layout()
        plt.show() # Zeige den eben erstellten Plot

        # Zusätzliche Plots für Skalen (z.B. 1-10)
        if scale_type == 'scale_1_10':
            # Histogramm
            plt.figure(figsize=(10, 5))
            hist_data = df_copy[q_col].dropna()
            sns.histplot(hist_data, bins=10, kde=False)
            plt.title(f'Histogramm für {q_col}')
            plt.xlabel(f"Antwort ({q_col})")
            plt.ylabel('Häufigkeit')
            min_val, max_val = hist_data.min(), hist_data.max()
            if pd.notna(min_val) and pd.notna(max_val):
                 plt.xticks(range(int(min_val), int(max_val) + 1))
            plt.show()

            # Boxplot
            plt.figure(figsize=(8, 3))
            sns.boxplot(x=hist_data)
            plt.title(f'Boxplot für {q_col}')
            plt.xlabel(f"Antwort ({q_col})")
            plt.show()

    # Gib den DataFrame zurück, falls Typkonvertierungen stattfanden
    # Wichtig: Wenn Konvertierungen stattfanden, muss die Variable im Hauptcode neu zugewiesen werden
    return df_copy

# --- Fragen analysieren ---

# Stelle sicher, dass df_clean existiert und korrekt geladen/bereinigt wurde
# z.B. df_clean = pd.read_csv(...)
# ... Code zum Bereinigen ...

# WICHTIG: df_clean nach jedem Aufruf neu zuweisen, falls Typkonvertierungen nötig waren!
# Q1: Family Importance
q1_labels = {1.0: 'Sehr wichtig', 2.0: 'Eher wichtig', 3.0: 'Nicht sehr wichtig', 4.0: 'Gar nicht wichtig'}
df_clean = analyze_question(df_clean, 'Q1', 'Wichtigkeit der Familie', labels=q1_labels)

# Q6: Religion Importance
q6_labels = {1.0: 'Sehr wichtig', 2.0: 'Eher wichtig', 3.0: 'Nicht sehr wichtig', 4.0: 'Gar nicht wichtig'}
df_clean = analyze_question(df_clean, 'Q6', 'Wichtigkeit der Religion', labels=q6_labels)

# Q18-Q26: Unwanted Neighbors (Binär: 1=Genannt, 2=Nicht genannt) in einem Grid
neighbor_labels = {1.0: 'Genannt', 2.0: 'Nicht genannt'}
neighbor_questions = {
    'Q18': 'Drogenabhängige', 'Q19': 'Menschen anderer Rasse', 'Q20': 'Menschen mit AIDS',
    'Q21': 'Einwanderer/Gastarbeiter', 'Q22': 'Homosexuelle', 'Q23': 'Menschen anderer Religion',
    'Q24': 'Starke Trinker', 'Q25': 'Unverheiratete Paare', 'Q26': 'Menschen anderer Sprache'
}

# Erstelle eine 3x3 Figur für die Nachbarschaftsfragen
fig, axes = plt.subplots(3, 3, figsize=(15, 12), sharey=True) # sharey für gleichen Maßstab auf y-Achse
axes = axes.flatten() # Macht die 2D-Array zu 1D für einfaches Iterieren

print("\n--- Analyse von Q18-Q26: Unerwünschte Nachbarn (1=Genannt, 2=Nicht genannt) ---")
# Iteriere durch die Fragen und die Achsen
for i, (q_col, q_desc) in enumerate(neighbor_questions.items()):
    # Rufe analyze_question für die entsprechende Achse auf
    # WICHTIG: Hier df_clean NICHT neu zuweisen, da die Funktion nur zum Plotten dient
    # und wir nicht wollen, dass die Typkonvertierung 9x im Original-df versucht wird.
    # Die Textausgabe erfolgt trotzdem für jede Frage.
    analyze_question(df_clean, q_col, q_desc, labels=neighbor_labels, scale_type='binary', ax=axes[i])

# Layout anpassen und Gesamtfigur anzeigen
fig.suptitle("Ablehnung verschiedener Gruppen als Nachbarn (1=Genannt, 2=Nicht genannt)", fontsize=16, y=1.0) # y etwas höher
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # rect anpassen, um Platz für suptitle zu lassen
plt.show()


# Q46: Happiness
q46_labels = {1.0: 'Sehr glücklich', 2.0: 'Eher glücklich', 3.0: 'Nicht sehr glücklich', 4.0: 'Gar nicht glücklich'}
df_clean = analyze_question(df_clean, 'Q46', 'Allgemeines Glücksempfinden', labels=q46_labels)

# Q71: Confidence in Government
q71_labels = {1.0: 'Sehr grosses Vertrauen', 2.0: 'Ziemlich viel Vertrauen', 3.0: 'Nicht sehr viel Vertrauen', 4.0: 'Überhaupt kein Vertrauen'}
df_clean = analyze_question(df_clean, 'Q71', 'Vertrauen in die Regierung', labels=q71_labels)

# Q94: Church Membership
q94_labels = {0.0: 'Kein Mitglied', 1.0: 'Inaktives Mitglied', 2.0: 'Aktives Mitglied'}
df_clean = analyze_question(df_clean, 'Q94', 'Mitgliedschaft Kirche/Religionsgemeinschaft', labels=q94_labels)

# Q163: Impact of Science & Technology (Scale 1-10)
# Hier wird die Funktion wieder normal aufgerufen (ax=None), um Einzelplots zu erzeugen
df_clean = analyze_question(df_clean, 'Q163', 'Auswirkung Wissenschaft/Technologie (1=Viel schlechter, 10=Viel besser)', scale_type='scale_1_10')

# Q190: Justification Parents Beating Children (Scale 1-10)
df_clean = analyze_question(df_clean, 'Q190', 'Rechtfertigung Eltern schlagen Kinder (1=Nie, 10=Immer)', scale_type='scale_1_10')

print("\n--- Ende der Analyse ausgewählter Q-Fragen ---")
```

Statistik=2.73: Der T-Wert. Er gibt an, wie gross der Unterschied zwischen den Mittelwerten im Verhältnis zur Streuung innerhalb der Gruppen ist. Ein grösserer (absoluter) T-Wert deutet auf einen deutlicheren Unterschied hin.

Der Unterschied ... ist statistisch signifikant (p < 0.05): Weil der p-Wert (0.006) kleiner ist als das übliche Signifikanzniveau (0.05 oder 5%), gehen wir davon aus, dass der beobachtete Unterschied nicht nur Zufall ist. Wir schliessen daraus, dass es einen "echten" (statistisch signifikanten) Unterschied in der durchschnittlichen Bewertung zwischen Männern und Frauen in der Grundgesamtheit gibt.
Welche Hypothese wurde getestet?
Der T-Test prüft die Nullhypothese (H0) gegen eine Alternativhypothese (H1).
H0 (Nullhypothese): Es gibt keinen Unterschied im mittleren Q163-Wert zwischen Männern und Frauen in der Grundgesamtheit. (Der Mittelwert ist für beide Gruppen gleich).
H1 (Alternativhypothese): Es gibt einen Unterschied im mittleren Q163-Wert zwischen Männern und Frauen in der Grundgesamtheit. (Die Mittelwerte sind nicht gleich).
Fazit: Da p < 0.05, verwerfen wir die Nullhypothese (H0). Wir haben statistische Evidenz dafür gefunden, dass sich die mittleren Bewertungen von Männern und Frauen unterscheiden.

**Wichtig: Statistische Signifikanz bedeutet nicht unbedingt praktische Relevanz.** Der Unterschied ist zwar statistisch nachweisbar, aber mit 7.30 vs. 7.21 ist er in der Praxis sehr klein.

2. Interpretation der ``Pearson Korrelation`` (Alter Q262 vs. Q163)
Was wurde gemessen?
Korrelationskoeffizient: 0.034: Misst die Stärke und Richtung eines linearen Zusammenhangs zwischen dem Alter (Q262) und der Bewertung (Q163). Ein Wert nahe 0 bedeutet einen sehr schwachen linearen Zusammenhang. Der positive Wert deutet einen minimalen Trend an: "höheres Alter geht mit minimal höherer Bewertung einher".
p-Wert: 0.000: (oft < 0.001). Die Wahrscheinlichkeit, eine Korrelation von 0.034 (oder stärker) zufällig zu beobachten, wenn es in Wirklichkeit gar keinen linearen Zusammenhang gäbe.
**Die Korrelation ist statistisch signifikant (p < 0.05):** Weil der p-Wert sehr klein ist, schliessen wir, dass die beobachtete Korrelation (auch wenn sie winzig ist) wahrscheinlich nicht genau Null ist in der Grundgesamtheit. Es gibt statistische Evidenz für einen (extrem schwachen) linearen Zusammenhang.
Widerspruch "signifikant" aber "sehr schwach"? Ein zentraler Punkt in der Statistik: Statistische Signifikanz ≠ Stärke des Effekts (oder praktische Relevanz).
Statistische Signifikanz sagt uns nur, ob ein Effekt (Unterschied, Zusammenhang) wahrscheinlich "echt" ist (also nicht nur Zufall). Bei sehr grossen Stichproben (wie dieser mit über 21'000 Fällen) können auch extrem kleine, praktisch unbedeutende Effekte statistisch signifikant werden. Das liegt daran, dass der Test sehr viel "Power" hat, um selbst minimale Abweichungen von der Nullhypothese (kein Zusammenhang) zu entdecken.
Stärke des Effekts wird durch den Korrelationskoeffizienten selbst (0.034) beschrieben. Dieser Wert sagt uns, wie stark der (lineare) Zusammenhang ist. Und 0.034 ist eben sehr, sehr schwach.

**Fazit:** Das Alter hat zwar einen statistisch nachweisbaren linearen Einfluss auf die Bewertung Q163, aber dieser Einfluss ist so gering, dass er praktisch kaum eine Rolle spielt. Das Wissen über das Alter einer Person hilft uns kaum dabei, ihre Antwort auf Q163 vorherzusagen (zumindest nicht linear).

3. ``Normalverteilung``
Die Normalverteilung wurde nicht explizit geprüft oder hergestellt.
T-Test: Ist relativ robust gegenüber Verletzungen der Normalverteilungsannahme, besonders bei grossen Stichproben wie dieser. Der Zentrale Grenzwertsatz sorgt dafür, dass die Verteilung der Mittelwerte sich einer Normalverteilung annähert, auch wenn die Rohdaten nicht normalverteilt sind. Daher ist **das Ergebnis des T-Tests hier wahrscheinlich trotzdem zuverlässig.**
**Pearson Korrelation: Setzt für die Signifikanzprüfung eigentlich auch Normalverteilung voraus. Aber auch hier gilt: Bei grossen Stichproben sind die Ergebnisse oft noch brauchbar.** Wichtiger ist aber: Pearson misst nur lineare Zusammenhänge. Wenn der Zusammenhang nicht-linear ist (z.B. U-förmig), wäre die Korrelation gering, obwohl ein Zusammenhang besteht. Das Streudiagramm hilft hier bei der Beurteilung.

1. Chi-Quadrat-Test: Land vs. Altersgruppe (Test b)
Dieser Test prüft, ob die Altersverteilung (Altersgruppe_Label) in verschiedenen Ländern (B_COUNTRY_ALPHA) unterschiedlich ist.
Nullhypothese (H0): Es gibt keinen Zusammenhang zwischen dem Land und der Altersgruppe. Die prozentuale Verteilung der Altersgruppen ist in allen betrachteten Ländern gleich.
Alternativhypothese (H1): Es gibt einen Zusammenhang. Die Altersstruktur unterscheidet sich zwischen den Ländern.
Wir führen den Test für die Top 15 Länder durch, um sicherzustellen, dass wir genügend Datenpunkte pro Zelle haben.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# --- Annahme: df_clean ist dein bereinigter DataFrame ---
# --- Stelle sicher, dass 'Altersgruppe_Label' existiert ---
if 'Altersgruppe_Label' not in df_clean.columns and 'X003R' in df_clean.columns:
    age_labels = {1.0: '16-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54', 5.0: '55-64', 6.0: '65+'}
    df_clean['Altersgruppe_Label'] = df_clean['X003R'].map(age_labels)
    print("Spalte 'Altersgruppe_Label' wurde erstellt.")
# --- Ende Label-Erstellung ---

print("\n--- Chi-Quadrat-Tests ---")

# Test b) Land vs. Altersgruppe
print("\nTest b) Land vs. Altersgruppe (Top 15 Länder):")
if 'B_COUNTRY_ALPHA' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    # Finde Top 15 Länder
    top_15_countries = df_clean['B_COUNTRY_ALPHA'].value_counts().nlargest(15).index.tolist()
    df_top15_countries = df_clean[df_clean['B_COUNTRY_ALPHA'].isin(top_15_countries)].dropna(subset=['B_COUNTRY_ALPHA', 'Altersgruppe_Label'])

    # Erstelle die Kontingenztafel (Kreuztabelle)
    contingency_table_country_age = pd.crosstab(df_top15_countries['B_COUNTRY_ALPHA'], df_top15_countries['Altersgruppe_Label'])
    print("Kontingenztafel (Land vs. Altersgruppe):")
    print(contingency_table_country_age)

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2, p, dof, expected = chi2_contingency(contingency_table_country_age)
        print(f"\nChi-Quadrat Statistik: {chi2:.2f}")
        print(f"Freiheitsgrade (dof): {dof}")
        print(f"p-Wert: {p:.3g}") # .3g für wissenschaftliche Notation bei sehr kleinen p-Werten

        if p < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p < 0.05). Die Altersstrukturen unterscheiden sich.")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p >= 0.05).")
            
        # Warnung bei niedrigen erwarteten Häufigkeiten (kann Test ungültig machen)
        if (expected < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5. Das Ergebnis des Chi-Quadrat-Tests könnte unzuverlässig sein.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected < 5).sum()}")


    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('B_COUNTRY_ALPHA', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

Unterschiedliche Altersstrukturen nach Land (Test b): Die Analyse der Teilnehmenden aus den Top-15-Ländern ergab einen signifikanten Zusammenhang zwischen dem Herkunftsland und der Altersgruppe (χ²(65) = 3354.05, p < 0.001). Dies bedeutet, dass sich die Altersverteilung der Befragten zwischen den untersuchten Ländern deutlich unterscheidet. Die Zusammensetzung der Stichprobe hinsichtlich junger, mittelalter und älterer Personen ist somit nicht über die Länder hinweg homogen.

Kruskal-Wallis-Test: Q190 nach Land

```python
from scipy.stats import kruskal

print("\n--- Kruskal-Wallis-Test: Q190 nach Land ---")

q190_col = 'Q190'
country_col = 'B_COUNTRY_ALPHA'

if q190_col in df_clean.columns and country_col in df_clean.columns:
    # Bereite Daten vor: Liste von Arrays, ein Array pro Land (mit genug Daten)
    df_test_kw = df_clean.dropna(subset=[q190_col, country_col])

    # Filtere Länder mit mindestens N gültigen Antworten (z.B. 30)
    min_samples = 30
    country_counts = df_test_kw[country_col].value_counts()
    valid_countries = country_counts[country_counts >= min_samples].index.tolist()

    data_per_country = [df_test_kw[df_test_kw[country_col] == country][q190_col].values for country in valid_countries]

    print(f"Führe Kruskal-Wallis für {len(valid_countries)} Länder mit >= {min_samples} Antworten für Q190 durch.")

    if len(data_per_country) > 1:
        try:
            stat_kw, p_kw = kruskal(*data_per_country) # Der Stern * entpackt die Liste
            print(f"\nKruskal-Wallis Statistik: {stat_kw:.2f}")
            print(f"p-Wert: {p_kw:.3g}")

            if p_kw < 0.05:
                print("-> Ergebnis: Es gibt einen statistisch signifikanten Unterschied in den Median-Bewertungen von Q190 zwischen den Ländern (p < 0.05).")
            else:
                print("-> Ergebnis: Es gibt keinen statistisch signifikanten Unterschied in den Median-Bewertungen von Q190 zwischen den Ländern (p >= 0.05).")

            # Visualisierung mit Boxplot (für die getesteten Länder)
            plt.figure(figsize=(15, 8))
            # Sortiere nach Median für bessere Übersicht
            df_plot_kw = df_test_kw[df_test_kw[country_col].isin(valid_countries)]
            median_order = df_plot_kw.groupby(country_col)[q190_col].median().sort_values().index
            sns.boxplot(x=country_col, y=q190_col, data=df_plot_kw, order=median_order, palette='coolwarm')
            plt.title(f'Verteilung der Q190-Bewertung nach Land (Länder mit >= {min_samples} Antworten)')
            plt.xlabel('Land')
            plt.ylabel('Q190 Bewertung (1=Nie, 10=Immer)')
            plt.xticks(rotation=90)
            plt.ylim(0, 11)
            plt.tight_layout()
            plt.show()

        except Exception as e:
            print(f"\nFehler beim Kruskal-Wallis-Test: {e}")
    else:
        print("\nNicht genügend Länder mit ausreichenden Daten für den Vergleich.")
else:
    print(f"\nBenötigte Spalten ({q190_col}, {country_col}) nicht vollständig vorhanden.")
```

**1. Was macht der Kruskal-Wallis-Test?**

*   Dieser Test ist eine Alternative zur ANOVA, wenn die Daten nicht normalverteilt sind (was bei Skalen wie 1-10 oft der Fall ist).
*   Er prüft, ob sich die **zentrale Tendenz** (genauer gesagt: der **Median**) einer Variable (hier Q190) zwischen **mehreren Gruppen** (hier den 14 Ländern) statistisch signifikant unterscheidet.
*   **Nullhypothese (H0):** Der Median der Q190-Bewertung ist in allen 14 verglichenen Ländern gleich. Es gibt keinen Unterschied in der "typischen" Antwort zwischen den Ländern.
*   **Alternativhypothese (H1):** Mindestens ein Land hat einen anderen Median-Wert für Q190 als die anderen Länder.

**2. Interpretation deiner Ergebnisse:**

*   `Führe Kruskal-Wallis für 14 Länder mit >= 30 Antworten für Q190 durch:` Der Test wurde nur für die 14 Länder durchgeführt, für die du mindestens 30 gültige Antworten auf Frage Q190 hattest. Das stellt sicher, dass die Ergebnisse pro Land nicht auf zu wenigen Meinungen basieren.
*   `Kruskal-Wallis Statistik: 3310.40:` Das ist der berechnete Wert des Tests (oft als H-Statistik bezeichnet). Je grösser dieser Wert, desto stärker unterscheiden sich die Rangsummen (und damit tendenziell die Mediane) zwischen den Gruppen.
*   `p-Wert: 0:` Das ist der wichtigste Wert. Er gibt die Wahrscheinlichkeit an, solche (oder noch grössere) Unterschiede in den Medianen zwischen den Ländern zufällig zu finden, *wenn eigentlich die Nullhypothese wahr wäre* (also wenn in Wirklichkeit alle Länder den gleichen Median hätten). Ein p-Wert von 0 (oder extrem nahe bei 0) bedeutet, dass es äusserst unwahrscheinlich ist, diese Unterschiede nur durch Zufall zu erhalten.
*   `Ergebnis: Es gibt einen statistisch signifikanten Unterschied ... (p < 0.05):` Weil der p-Wert (0) kleiner als das übliche Signifikanzniveau von 0.05 ist, **verwerfen wir die Nullhypothese**. Das bedeutet: Der Test liefert starke statistische Evidenz dafür, dass sich die "typische" Antwort auf die Frage Q190 (Rechtfertigung von Kinderschlägen) zwischen den 14 untersuchten Ländern **tatsächlich unterscheidet**.

**3. Verbindung zum Boxplot (dem angehängten Bild):**

Chi-Quadrat-Test: Q190 (kategorisiert) vs. Altersgruppe

```python
from scipy.stats import chi2_contingency

print("\n--- Chi-Quadrat-Test: Q190 (kategorisiert) vs. Altersgruppe ---")

if 'Q190_Kat' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    df_test_chi_q190_age = df_clean.dropna(subset=['Q190_Kat', 'Altersgruppe_Label'])

    contingency_q190_age = pd.crosstab(df_test_chi_q190_age['Q190_Kat'], df_test_chi_q190_age['Altersgruppe_Label'])
    print("Kontingenztafel (Q190 Kategorie vs. Altersgruppe):")
    print(contingency_q190_age)

    # Visualisierung als gestapeltes Balkendiagramm (Prozente)
    cont_q190_age_perc = contingency_q190_age.apply(lambda x: x*100/sum(x), axis=0)
    print("\nProzentuale Verteilung pro Altersgruppe:")
    print(cont_q190_age_perc.round(1))

    cont_q190_age_perc.T.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='coolwarm', rot=0)
    plt.title('Verteilung der Q190-Bewertung (kategorisiert) nach Altersgruppe')
    plt.xlabel('Altersgruppe')
    plt.ylabel('Prozent (%)')
    plt.legend(title='Q190 Kategorie', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2_qa, p_qa, dof_qa, expected_qa = chi2_contingency(contingency_q190_age)
        print(f"\nChi-Quadrat Statistik: {chi2_qa:.2f}")
        print(f"Freiheitsgrade (dof): {dof_qa}")
        print(f"p-Wert: {p_qa:.3g}")

        if p_qa < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p < 0.05).")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p >= 0.05).")

        if (expected_qa < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected_qa < 5).sum()}")

    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('Q190_Kat', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

*   **Deine Skalen (1-10):** Bei Skalen wie 1-10 (z.B. Q163, Q190) ist eine perfekte Normalverteilung eher unwahrscheinlich. Oft sind sie "ordinal" (haben eine Reihenfolge) aber nicht "intervallskaliert" im strengen Sinne (der Abstand zwischen 1 und 2 ist nicht zwingend gleich dem Abstand zwischen 9 und 10). Die Verteilungen sind oft schief oder haben Peaks an den Enden.
*   **Grosse Stichproben:** Bei sehr grossen Stichproben wie deiner (N > 21.000) greift der **Zentrale Grenzwertsatz**. Dieser besagt, dass sich die *Verteilung der Mittelwerte* aus vielen Stichproben einer Normalverteilung annähert, selbst wenn die ursprüngliche Variable nicht normalverteilt ist. Das macht viele statistische Tests (wie den t-Test oder die lineare Regression) relativ robust gegenüber Verletzungen der Normalverteilungsannahme, *wenn die Stichprobe gross genug ist*. Die Signifikanztests sind also meist trotzdem gültig.
*   **Ist es "erledigt"?** Für die **deskriptive Phase:** Ja, wir haben die Verteilungen wichtiger Variablen beschrieben. Wir wissen, dass z.B. Q163 und Q190 nicht perfekt normalverteilt sind. Für die **inferenzstatistische Phase** (wie die Tests, die wir gemacht haben): Wir haben uns (implizit) auf die Robustheit der Tests bei grossen Stichproben verlassen oder (wie beim Kruskal-Wallis-Test) eine Methode gewählt, die keine Normalverteilung voraussetzt. Man *könnte* für jede Variable formale Tests auf Normalverteilung (z.B. Kolmogorov-Smirnov, Shapiro-Wilk) durchführen, aber bei N>21000 werden diese fast immer signifikant ausfallen (also "nicht normalverteilt" anzeigen), selbst bei leichten Abweichungen. Aussagekräftiger ist oft die visuelle Inspektion (Histogramm) und die Prüfung der Voraussetzungen für den *konkreten* Test, den man anwenden will.

**Fazit zur Normalverteilung:** Du musst nicht den ganzen Datensatz auf Normalverteilung prüfen. Wir haben die Verteilungen wichtiger Variablen deskriptiv erfasst. Für die angewendeten Tests ist die Annahme entweder nicht zwingend erforderlich oder die Tests sind aufgrund der Stichprobengrösse robust. Du hast hier nichts Wichtiges übersehen.

**2. Sache: Weitere Analysen (Regression, PCA, Clustering)**

Das sind spannende Ideen, um tiefer in die Daten einzusteigen!

*   **Multiple lineare Regression (MLR):**
    *   **Ziel:** Vorhersage einer abhängigen (metrischen) Variable durch mehrere unabhängige Variablen.
    *   **Anwendung bei dir:** Wir könnten versuchen, `Q163` (Bewertung Wiss./Technik) vorherzusagen.
    *   **Mögliche unabhängige Variablen:** Demografie (`Q262`=Alter, `Q260`=Geschlecht, evtl. Bildung?), andere Einstellungen (z.B. `Q190`? Oder andere technologiebezogene Fragen?), `B_COUNTRY_ALPHA` (als Dummy-Variablen).
    *   **Vorgehen:**
        1.  Auswahl der abhängigen und unabhängigen Variablen.
        2.  Prüfung der Voraussetzungen der MLR (Linearität, Unabhängigkeit der Fehler, Homoskedastizität, Normalverteilung der *Residuen* - nicht der Originalvariablen!).
        3.  Modellberechnung und Interpretation (Koeffizienten, R-Quadrat).

*   **Vorschlag für weiteres Verfahren: Clusteranalyse (z.B. K-Means)**
    *   **Ziel:** Gruppen (Cluster) von ähnlichen Beobachtungen (hier: Befragten) auf Basis mehrerer Variablen zu identifizieren, ohne vorher zu wissen, welche Gruppen es gibt.
    *   **Anwendung bei dir (Idee 2):** Das passt **perfekt** zu deiner Frage nach Mustern bei Personen, die Kinderschläge stark rechtfertigen (Q190 = 8-10)!
    *   **Vorgehen:**
        1.  **Subset erstellen:** Filtere den Datensatz auf Befragte mit `Q190 >= 8`.
        2.  **Variablen auswählen:** Wähle Variablen aus, die das Profil dieser Personen charakterisieren könnten (z.B. Fragen zu Religiosität, sozialen Normen wie Homosexualität, Vertrauen in Institutionen, Demografie wie Alter, Bildung, Land).
        3.  **Daten vorbereiten:** Numerische Variablen standardisieren (auf gleichen Massstab bringen).
        4.  **Clustering-Algorithmus anwenden:** Z.B. K-Means. Man muss die Anzahl der Cluster (k) wählen (z.B. durch Ausprobieren oder die "Ellbogenmethode").
        5.  **Cluster interpretieren:** Analysiere die resultierenden Cluster, indem du die Durchschnittswerte der Profil-Variablen für jedes Cluster vergleichst. Gibt es z.B. ein Cluster, das durch hohe Religiosität und niedriges Alter gekennzeichnet ist?

*   **Principal Component Analysis (PCA):**
    *   **Ziel:** Dimensionalitätsreduktion. Fasst viele (korrelierte) Variablen zu wenigen neuen, unkorrelierten Variablen ("Hauptkomponenten" oder "Principal Components") zusammen, die möglichst viel der ursprünglichen Varianz erklären.
    *   **Anwendung bei dir (Idee 1):** Passt gut, um herauszufinden, welche *latenten Dimensionen* in deinen 20-30 ausgewählten Fragen stecken und wie diese mit Q163 zusammenhängen.
    *   **Vorgehen:**
        1.  **Variablen auswählen:** Wähle die 20-30 Fragen aus, von denen du annimmst, dass sie mit Q163 zusammenhängen könnten. Wichtig: Sie sollten tendenziell metrisch sein oder zumindest eine klare Ordnung haben.
        2.  **Daten vorbereiten:** Nur Befragte ohne fehlende Werte in diesen Variablen verwenden. Variablen standardisieren (sehr wichtig für PCA!).
        3.  **PCA durchführen:** Berechne die Hauptkomponenten.
        4.  **Komponenten interpretieren:** Schau dir an, wie viel Varianz jede Komponente erklärt (Eigenwerte) und welche Originalvariablen stark auf welche Komponente "laden" (Komponentenladungen). Das gibt den Komponenten ihre Bedeutung (z.B. "Technikoptimismus", "Soziale Normen").
        5.  **Zusammenhang mit Q163:** Berechne die Korrelation zwischen den Scores der wichtigsten Hauptkomponenten und der Variable Q163.

**Zusammenfassend zu den Verfahren:**

*   **MLR:** Gut, um eine spezifische Variable (wie Q163) zu *erklären/vorherzusagen*.
*   **PCA:** Gut, um die *Struktur* in einer grösseren Menge von Variablen aufzudecken und Daten zu reduzieren.
*   **Clusteranalyse:** Gut, um *Gruppen von ähnlichen Personen* zu finden, basierend auf ihren Antworten über mehrere Variablen hinweg.

Alle drei Verfahren sind sinnvoll und würden deine Analyse deutlich über die deskriptive Ebene hinausheben.

# PCA

## Schritt 1: Themenbereichen für die Analyse
Über alle 300 Fragen eine PCA zu machen wäre suboptimal

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer # Werden wir evtl. brauchen
from sklearn.decomposition import PCA # Für PCA
import matplotlib.pyplot as plt
import seaborn as sns

# --- Annahme: df_clean ist dein bereinigter DataFrame ---

# 1. Variablen auswählen
pca_block1_vars = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6']

# Stelle sicher, dass alle Variablen existieren
missing_pca_vars = [var for var in pca_block1_vars if var not in df_clean.columns]
if missing_pca_vars:
    print(f"Fehler: Folgende Variablen für PCA Block 1 wurden nicht gefunden: {missing_pca_vars}")
    # Stopp
else:
    df_pca1 = df_clean[pca_block1_vars].copy()
    print(f"DataFrame für PCA Block 1 mit {len(df_pca1.columns)} Variablen erstellt.")
    initial_rows_pca1 = len(df_pca1)
    print(f"Ursprüngliche Anzahl Zeilen: {initial_rows_pca1}")

    # 2. Fehlende Werte prüfen
    print("\n--- Fehlende Werte (PCA Block 1) ---")
    print(df_pca1.isnull().sum())
    print(f"Anzahl Zeilen mit mindestens einem NaN: {df_pca1.isnull().any(axis=1).sum()}")

    # Behandeln fehlender Werte: Listwise Deletion (Alternative wäre Imputation)
    df_pca1.dropna(inplace=True)
    rows_after_dropna_pca1 = len(df_pca1)
    print(f"\n{initial_rows_pca1 - rows_after_dropna_pca1} Zeilen wurden wegen fehlender Werte entfernt.")
    print(f"Verbleibende Zeilen für die PCA: {rows_after_dropna_pca1}")

    if rows_after_dropna_pca1 < len(pca_block1_vars) * 10: # Faustregel: Mindestens 10 Fälle pro Variable
         print("Warnung: Nach Entfernen fehlender Werte sind relativ wenige Fälle übrig. PCA könnte instabil sein.")
    elif rows_after_dropna_pca1 > 0:
        print("Fortfahren mit Datenvorbereitung...")

        # 3. Variablen umkodieren (höher = wichtiger)
        # Skala: 1=Sehr wichtig, 2=Eher wichtig, 3=Nicht sehr wichtig, 4=Gar nicht wichtig
        # Umkehrung: 1->4, 2->3, 3->2, 4->1
        print("\n--- Variablen umkodieren (höher = wichtiger) ---")
        for var in pca_block1_vars:
             if var in df_pca1.columns:
                 df_pca1[var] = df_pca1[var].replace({1: 4, 2: 3, 3: 2, 4: 1})
                 print(f"Variable {var} umkodiert.")

        # Stelle sicher, dass alles numerisch ist
        try:
             df_pca1 = df_pca1.astype(float)
        except Exception as e:
             print(f"Fehler bei Konvertierung zu numerischem Typ: {e}")
             # Stopp hier, wenn Fehler auftritt

        # 4. Daten standardisieren (Z-Transformation: Mittelwert=0, StdAbw=1) - WICHTIG für PCA!
        print("\n--- Daten standardisieren ---")
        scaler_pca1 = StandardScaler()
        df_pca1_scaled = scaler_pca1.fit_transform(df_pca1)

        # Konvertiere zurück zu DataFrame für bessere Handhabung (optional)
        df_pca1_scaled = pd.DataFrame(df_pca1_scaled, index=df_pca1.index, columns=df_pca1.columns)
        print("Daten erfolgreich standardisiert.")
        print("Beispiel der skalierten Daten (erste 5 Zeilen):")
        print(df_pca1_scaled.head())

    else:
        print("Keine Daten übrig nach Entfernen von NaN. PCA nicht möglich.")
```

```python
# Nur weitermachen, wenn df_pca1_scaled existiert und nicht leer ist
if 'df_pca1_scaled' in locals() and not df_pca1_scaled.empty:
    print("\n--- Durchführung der PCA (Block 1: Lebensprioritäten) ---")

    # Initialisiere PCA ohne Angabe der Komponentenanzahl
    # n_components=None bedeutet, dass alle Komponenten berechnet werden
    pca = PCA(n_components=None)
    pca.fit(df_pca1_scaled)

    # Ergebnisse analysieren
    explained_variance_ratio = pca.explained_variance_ratio_
    cumulative_variance_ratio = np.cumsum(explained_variance_ratio)

    print("\nErklärte Varianz pro Komponente:")
    for i, ratio in enumerate(explained_variance_ratio):
        print(f"  PC{i+1}: {ratio:.4f} ({(ratio*100):.1f}%)")

    print("\nKumulierte erklärte Varianz:")
    for i, ratio in enumerate(cumulative_variance_ratio):
        print(f"  Bis PC{i+1}: {ratio:.4f} ({(ratio*100):.1f}%)")

    # --- Scree Plot ---
    plt.figure(figsize=(10, 6))
    num_components = len(explained_variance_ratio)
    plt.plot(range(1, num_components + 1), explained_variance_ratio, marker='o', linestyle='--', label='Erklärte Varianz pro Komponente')
    plt.plot(range(1, num_components + 1), cumulative_variance_ratio, marker='s', linestyle='-', label='Kumulierte erklärte Varianz')

    # Optional: Kaiser-Kriterium (Eigenwert > 1) - Eigenwerte sind pca.explained_variance_
    # plt.axhline(y=1/num_components, color='r', linestyle=':', label='Grenze Eigenwert=1 (approx.)') # Näherung

    plt.title('Scree Plot für PCA Block 1 (Lebensprioritäten)')
    plt.xlabel('Anzahl der Hauptkomponenten')
    plt.ylabel('Anteil der erklärten Varianz')
    plt.xticks(range(1, num_components + 1))
    plt.legend()
    plt.grid(True)
    plt.show()

    print("\nEntscheidung über Anzahl der Komponenten:")
    print(" - Kaiser-Kriterium: Behalte Komponenten mit Eigenwert > 1 (entspricht erklärter Varianz > 1/Anzahl Variablen).")
    print(" - Ellbogen-Kriterium (Scree Plot): Suche den 'Knick' in der Kurve der erklärten Varianz pro Komponente.")
    print(" - Prozent der erklärten Varianz: Wähle so viele Komponenten, dass ein ausreichender Prozentsatz (z.B. 60-80%) der Gesamtvarianz erklärt wird.")

else:
    print("Keine standardisierten Daten vorhanden, PCA nicht möglich.")
```

```python
# Nur weitermachen, wenn pca-Objekt existiert und wir uns für k entschieden haben
if 'pca' in locals() and 'df_pca1' in locals(): # df_pca1 hat die ursprünglichen Spaltennamen
    num_components_to_keep = 3
    print(f"\n--- Interpretation der ersten {num_components_to_keep} Hauptkomponenten ---")

    # Lade die Komponenten-Ladungen (Eigenvektoren skaliert mit Wurzel der Eigenwerte)
    # pca.components_ sind die Eigenvektoren (Zeilen=Komponenten, Spalten=Variablen)
    loadings = pca.components_[:num_components_to_keep, :] # Wähle nur die ersten k Komponenten

    # Erstelle einen DataFrame für die Ladungen
    loadings_df = pd.DataFrame(loadings.T, # Transponieren für Variablen als Zeilen
                              columns=[f'PC{i+1}' for i in range(num_components_to_keep)],
                              index=df_pca1.columns) # Original-Variablennamen als Index

    print("\nKomponenten-Ladungen:")
    print(loadings_df.round(3))

    # Visualisierung der Ladungen (Heatmap)
    plt.figure(figsize=(8, 6))
    sns.heatmap(loadings_df, annot=True, cmap='coolwarm', fmt=".3f", center=0)
    plt.title(f'Ladungen der Variablen auf die ersten {num_components_to_keep} Hauptkomponenten')
    plt.xlabel('Hauptkomponente')
    plt.ylabel('Originalvariable')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

    print("\nInterpretation der Ladungen:")
    print(" - Eine hohe *positive* Ladung bedeutet, dass hohe Werte der Variable stark mit hohen Werten der Komponente assoziiert sind.")
    print(" - Eine hohe *negative* Ladung bedeutet, dass hohe Werte der Variable stark mit *niedrigen* Werten der Komponente assoziiert sind.")
    print(" - Werte nahe Null bedeuten, dass die Variable wenig mit der Komponente zu tun hat.")
    print("Versuche, jeder Komponente einen Namen zu geben, basierend auf den Variablen, die auf sie am stärksten (positiv oder negativ) laden.")

else:
    print("PCA-Objekt nicht gefunden, Interpretation nicht möglich.")
```

```python
# Nur weitermachen, wenn df_mlr existiert und gültig ist und IVs definiert sind
if 'df_mlr' in locals() and not df_mlr.empty and all(v in df_mlr.columns for v in ivs) and dv in df_mlr.columns:

    print("\n--- Berechnung des Multiplen Linearen Regressionsmodells ---")

    # Definiere abhängige (y) und unabhängige (X) Variablen
    y = df_mlr[dv]
    X = df_mlr[ivs]

    # Füge eine Konstante (Intercept) zum Modell hinzu
    # Das ist notwendig, damit das Modell den Achsenabschnitt schätzt
    X = sm.add_constant(X)

    # Berechne das OLS (Ordinary Least Squares) Modell
    model = sm.OLS(y, X)
    results = model.fit()

    # Gib die Zusammenfassung der Regressionsergebnisse aus
    print("\nRegressionsergebnisse:")
    print(results.summary())

    # --- Residuenanalyse (vereinfacht) ---
    print("\n--- Residuenanalyse (zur Prüfung einiger Annahmen) ---")
    residuals = results.resid
    fitted_values = results.fittedvalues

    # 1. Histogramm der Residuen (sollten annähernd normalverteilt sein)
    plt.figure(figsize=(10, 5))
    sns.histplot(residuals, kde=True)
    plt.title('Histogramm der Residuen')
    plt.xlabel('Residuen (Fehler)')
    plt.ylabel('Häufigkeit')
    plt.show()
    print("Die Residuen sollten idealerweise einer Normalverteilung ähneln (Glockenkurve).")

    # 2. Streudiagramm: Angepasste Werte vs. Residuen (Homoskedastizität)
    plt.figure(figsize=(10, 5))
    sns.scatterplot(x=fitted_values, y=residuals)
    plt.axhline(0, color='red', linestyle='--') # Füge Nulllinie hinzu
    plt.title('Angepasste Werte vs. Residuen')
    plt.xlabel('Angepasste Werte (Vorhersagen)')
    plt.ylabel('Residuen (Fehler)')
    plt.show()
    print("Die Punkte sollten zufällig um die Nulllinie streuen, ohne erkennbares Muster (z.B. Trichterform).")
    print("Muster deuten auf Heteroskedastizität hin (Verletzung einer Annahme).")


else:
    print("Voraussetzungen für Regressionsberechnung nicht erfüllt.")
```

## Korrelationsanalyse, da MLR versagt

Nein, **für die Berechnung der Korrelationen selbst musst du die Variablen nicht standardisieren.**

Hier ist der Grund:

*   Der **Pearson-Korrelationskoeffizient (r)** ist von Natur aus **skaleninvariant**. Er misst die Stärke und Richtung des *linearen* Zusammenhangs zwischen zwei Variablen, unabhängig davon, in welchen Einheiten oder auf welchen Skalen (z.B. 1-4, 1-10, 0-100) diese Variablen gemessen werden.
*   Die Formel für die Korrelation berücksichtigt die Mittelwerte und Standardabweichungen der Variablen bereits intern, sodass das Ergebnis immer zwischen -1 und +1 liegt.
*   Wenn du die Variablen vorher standardisieren würdest (Mittelwert 0, Standardabweichung 1), würdest du **exakt dieselben Korrelationskoeffizienten** erhalten wie ohne Standardisierung.

**Wofür ist Standardisierung wichtig?**

Standardisierung ist entscheidend für Verfahren, bei denen die *tatsächlichen Werte* oder die *Distanzen* zwischen Punkten eine Rolle spielen und die Ergebnisse durch unterschiedliche Skalen beeinflusst werden könnten, wie z.B.:

*   **K-Means Clustering** (haben wir gemacht)
*   **PCA (Hauptkomponentenanalyse)** (hatten wir diskutiert)

```python
import pandas as pd
import numpy as np

# --- Annahme: df_clean ist dein bereinigter DataFrame ---

target_var = 'Q163'

# 1. Identifiziere alle Q-Variablen
q_vars = [col for col in df_clean.columns if col.startswith('Q')]

if target_var not in q_vars:
    print(f"Fehler: Zielvariable {target_var} nicht im DataFrame gefunden.")
    # Stopp
elif len(q_vars) <= 1:
    print("Fehler: Keine anderen Q-Variablen zum Korrelieren gefunden.")
    # Stopp
else:
    print(f"Analysiere Korrelationen zwischen {target_var} und {len(q_vars) - 1} anderen Q-Variablen.")

    # 2. Wähle relevante Spalten und stelle sicher, dass sie numerisch sind
    df_corr = df_clean[[target_var] + [var for var in q_vars if var != target_var]].copy()

    # Versuche, alle Spalten in numerisch umzuwandeln (ignoriere Fehler für nicht-konvertierbare)
    for col in df_corr.columns:
        df_corr[col] = pd.to_numeric(df_corr[col], errors='coerce')

    # Entferne Spalten, die nicht numerisch wurden (falls vorhanden)
    non_numeric_cols = df_corr.select_dtypes(exclude=[np.number]).columns.tolist()
    if non_numeric_cols:
        print(f"Warnung: Folgende Q-Spalten sind nicht numerisch und werden ignoriert: {non_numeric_cols}")
        df_corr = df_corr.select_dtypes(include=[np.number])

    # 3. Berechne Korrelationen
    # .corr() berechnet paarweise Korrelationen und ignoriert NaNs (pairwise deletion)
    correlations = df_corr.corr()[target_var]

    # Entferne die Korrelation der Variable mit sich selbst (wäre 1)
    correlations = correlations.drop(target_var, errors='ignore')

    # 4. Ergebnisse anzeigen
    print(f"\nKorrelationen aller Q-Variablen mit {target_var}:")

    # Sortiere nach dem absoluten Wert der Korrelation (absteigend)
    sorted_correlations = correlations.abs().sort_values(ascending=False)

    # Zeige z.B. die Top 20 stärksten Korrelationen
    num_to_show = 20
    print(f"\nTop {num_to_show} Variablen nach stärkster *absoluter* Korrelation mit {target_var}:")
    # Füge den ursprünglichen Korrelationswert (mit Vorzeichen) hinzu
    top_corr_df = pd.DataFrame({
        'Correlation': correlations.loc[sorted_correlations.head(num_to_show).index],
        'Absolute Correlation': sorted_correlations.head(num_to_show)
    })
    print(top_corr_df)

    # Optional: Separate Listen für positive und negative Korrelationen
    positive_corr = correlations[correlations > 0].sort_values(ascending=False)
    negative_corr = correlations[correlations < 0].sort_values(ascending=True) # Ascending for most negative

    print(f"\nTop 10 *positiv* korrelierenden Variablen mit {target_var}:")
    print(positive_corr.head(10))

    print(f"\nTop 10 *negativ* korrelierenden Variablen mit {target_var}:")
    print(negative_corr.head(10))
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import seaborn as sns
import matplotlib.pyplot as plt

# --- Annahme: df_clean ist dein bereinigter DataFrame ---

# 1. Variablen auswählen
dv_new = 'Q163'
ivs_new = [
    'Q246', 'Q158', 'Q48', # Positive
    'Q44', 'Q76', 'Q253', # Negative
    'Q262', # Alter
    'Q260' # Geschlecht (wird zu Dummy 'Female')
]
all_vars_new = [dv_new] + ivs_new

# Stelle sicher, dass alle Variablen existieren
missing_vars_new = [var for var in all_vars_new if var not in df_clean.columns]
if missing_vars_new:
    print(f"Fehler: Folgende Variablen für das neue MLR-Modell wurden nicht gefunden: {missing_vars_new}")
    # Stopp
else:
    df_mlr_new = df_clean[all_vars_new].copy()
    print(f"DataFrame für neues MLR mit {len(df_mlr_new.columns)} Variablen erstellt.")

    # 2. Fehlende Werte behandeln (Listwise Deletion)
    print("\n--- Fehlende Werte (Neues Modell) ---")
    print("Anzahl fehlender Werte pro Variable (vorher):")
    print(df_mlr_new.isnull().sum())
    initial_rows_new = len(df_mlr_new)
    df_mlr_new.dropna(inplace=True)
    rows_after_dropna_new = len(df_mlr_new)
    print(f"\n{initial_rows_new - rows_after_dropna_new} Zeilen wurden wegen fehlender Werte entfernt.")
    print(f"Verbleibende Zeilen für die neue Regression: {rows_after_dropna_new}")

    if rows_after_dropna_new < (len(ivs_new) + 50):
         print("Warnung: Nach Entfernen fehlender Werte sind relativ wenige Fälle übrig.")
    elif rows_after_dropna_new > 0:
        print("Fortfahren mit Datenvorbereitung...")

        # 3. Dummy-Variable für Geschlecht (Q260)
        final_iv_list = list(ivs_new) # Kopiere die Liste für finale IVs
        if 'Q260' in df_mlr_new.columns:
            df_mlr_new['Female'] = (df_mlr_new['Q260'] == 2).astype(int)
            final_iv_list.remove('Q260')
            final_iv_list.append('Female')
            df_mlr_new.drop('Q260', axis=1, inplace=True)
            print("\nDummy-Variable 'Female' erstellt.")
        else:
             print("Warnung: Q260 nicht im DataFrame.")
             if 'Q260' in final_iv_list: final_iv_list.remove('Q260') # Entferne aus IV-Liste


        # Stelle sicher, dass alle verbleibenden Spalten numerisch sind
        try:
            df_mlr_new = df_mlr_new.astype(float)
            print("\nAlle Variablen sind numerisch.")

            # 4. Prüfung auf Multikollinearität (Neues Modell)
            print("\n--- Prüfung auf Multikollinearität (Neues Modell) ---")
            X_new = df_mlr_new[final_iv_list]
            corr_matrix_new = X_new.corr()

            print("Korrelationsmatrix (Neues Modell):")
            print(corr_matrix_new.round(2))

            plt.figure(figsize=(9, 7))
            sns.heatmap(corr_matrix_new, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
            plt.title('Korrelationsmatrix der unabhängigen Variablen (Neues Modell)')
            plt.show()

            # VIF
            X_vif_new = sm.add_constant(X_new)
            vif_data_new = pd.DataFrame()
            vif_data_new["Variable"] = X_vif_new.columns
            vif_data_new["VIF"] = [variance_inflation_factor(X_vif_new.values, i) for i in range(X_vif_new.shape[1])]
            vif_data_new = vif_data_new[vif_data_new["Variable"] != 'const']
            print("\nVariance Inflation Factor (VIF) (Neues Modell):")
            print(vif_data_new.round(2))

            # Prüfe auf hohe VIF-Werte
            if (vif_data_new['VIF'] >= 5).any():
                 print("\nWarnung: Mindestens eine Variable hat einen VIF >= 5. Multikollinearität könnte ein Problem sein.")
            else:
                 print("\nKeine Anzeichen für problematische Multikollinearität (alle VIF < 5).")

        except Exception as e:
            print(f"\nFehler bei Datenvorbereitung oder Multikollinearitätsprüfung: {e}")

    else:
        print("Keine Daten übrig nach Entfernen von NaN. Regression nicht möglich.")
```
