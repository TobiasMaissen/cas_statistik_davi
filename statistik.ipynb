## Schritt 1: Datenbereinigung und -aufbereitung

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Annahme: df_2020_final ist Ihr DataFrame
df_clean = df_2020_final.copy()

# Liste der Spalten, in denen negative Werte NaN bedeuten könnten
# (Passen Sie diese Liste ggf. an, basierend auf Ihrem Codebuch)
cols_to_clean = [col for col in df_clean.columns if col.startswith('Q')]
cols_to_clean.extend(['X003R']) # Fügen Sie weitere demografische Spalten hinzu, falls nötig

# Ersetzen Sie negative Werte durch NaN
for col in cols_to_clean:
    # Sicherstellen, dass die Spalte numerisch ist, bevor wir vergleichen
    if pd.api.types.is_numeric_dtype(df_clean[col]):
        df_clean.loc[df_clean[col] < 0, col] = np.nan
    else:
        # Versuch, die Spalte in numerisch umzuwandeln, falls sie als Objekt gelesen wurde
        try:
            numeric_col = pd.to_numeric(df_clean[col], errors='coerce')
            df_clean[col] = numeric_col # Originalspalte überschreiben
            df_clean.loc[df_clean[col] < 0, col] = np.nan
            print(f"Spalte {col} wurde in numerisch konvertiert und bereinigt.")
        except Exception as e:
            print(f"Spalte {col} konnte nicht sicher bereinigt werden: {e}")


# Spezifische Bereinigung für Q260 (Geschlecht), falls nötig (Codebuch prüfen!)
if 'Q260' in df_clean.columns and pd.api.types.is_numeric_dtype(df_clean['Q260']):
     df_clean.loc[df_clean['Q260'] < 0, 'Q260'] = np.nan
```

**2. QQ-Plots, t-Tests, Chi-Quadrat-Tests etc.**
Deine Verwirrung ist absolut verständlich, diese statistischen Tests können am Anfang überwältigend sein. Lass uns das kurz einordnen:

- **Was wir gerade machen:** Deskriptive Statistik. Wir beschreiben, wie die Daten aussehen (Häufigkeiten, Mittelwerte, Streuungen, Verteilungen durch Histogramme/Boxplots). Ziel ist es, ein Gefühl für die Daten zu bekommen.

- **Was diese Tests machen:** Inferenzstatistik. Sie werden verwendet, um Hypothesen zu testen oder Rückschlüsse von der Stichprobe auf die Grundgesamtheit zu ziehen.
  - **QQ-Plot:** Ein visuelles Werkzeug, um zu prüfen, ob Daten einer bestimmten theoretischen Verteilung folgen (oft Normalverteilung). Nützlich nachdem man sich die Verteilung z.B. im Histogramm angesehen hat, um die Abweichung von der Normalität genauer zu beurteilen. Könnten wir z.B. für Q262 (Alter) machen, um die Abweichung von der Normalverteilung deutlicher zu sehen.
  - **t-Test:** Vergleicht die Mittelwerte einer (ungefähr) normalverteilten Variable zwischen zwei Gruppen. Beispiel: "Ist der durchschnittliche Q163-Score bei Männern signifikant anders als bei Frauen?" (Voraussetzungen wie Normalverteilung innerhalb der Gruppen und Varianzhomogenität müssen geprüft werden, bei grossen Stichproben ist der Test aber robust gegen Verletzungen der Normalverteilungsannahme).
  - **Chi-Quadrat-Test (χ²):** Prüft, ob ein Zusammenhang zwischen zwei kategorialen Variablen besteht. Beispiel: "Gibt es einen Zusammenhang zwischen der Altersgruppe (X003R) und dem Land (B_COUNTRY_ALPHA)?" oder "Gibt es einen Zusammenhang zwischen Geschlecht (Q260) und einer kategorialen Antwort auf eine Frage?".
  
Wann verwenden? Diese Tests kommen meist nach der deskriptiven Analyse ins Spiel, wenn du spezifische Fragen oder Hypothesen hast, die du testen möchtest.
Beispiel: Nachdem wir sehen, dass Männer und Frauen leicht unterschiedliche Durchschnittswerte bei Q163 haben (deskriptiv), könnten wir einen t-Test verwenden, um zu prüfen, ob dieser Unterschied statistisch signifikant ist (inferenziell).
Beispiel: Nachdem wir sehen, dass die Verteilung der Altersgruppen je nach Land unterschiedlich aussieht (deskriptiv in Kreuztabellen), könnten wir einen Chi-Quadrat-Test verwenden, um zu prüfen, ob dieser Zusammenhang signifikant ist (inferenziell).

```python
# --- Analyse: Altersgruppenverteilung pro Land (Heatmap) ---
print("\n--- Analyse: Altersgruppenverteilung pro Land (Heatmap) ---")

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np # Für np.nan wenn Daten fehlen

# Sicherstellen, dass df_clean und die Spalten existieren
if 'df_clean' not in locals() or 'X003R' not in df_clean.columns or 'B_COUNTRY_ALPHA' not in df_clean.columns:
    print("Fehler: Benötigte Daten (df_clean, X003R, B_COUNTRY_ALPHA) nicht gefunden.")
else:
    # Mapping und Reihenfolge für Altersgruppen-Labels
    age_labels = {1.0: '16-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54', 5.0: '55-64', 6.0: '65+'}
    ordered_age_labels = list(age_labels.values())

    # Daten vorbereiten (wie zuvor)
    df_age_country = df_clean[['B_COUNTRY_ALPHA', 'X003R']].dropna()
    df_age_country = df_age_country[df_age_country['X003R'].isin(age_labels.keys())]
    df_age_country['Altersgruppe_Label'] = df_age_country['X003R'].map(age_labels)

    # Kreuztabelle mit absoluten Zahlen
    crosstab_abs = pd.crosstab(df_age_country['B_COUNTRY_ALPHA'], df_age_country['Altersgruppe_Label'])
    crosstab_abs = crosstab_abs.reindex(columns=ordered_age_labels, fill_value=0)

    # Kreuztabelle in Prozent umrechnen
    # Füge kleine Konstante hinzu, um Division durch Null bei Ländern ohne Antworten zu vermeiden
    row_sums = crosstab_abs.sum(axis=1)
    # Länder ohne Antworten (Summe=0) identifizieren und für diese Zeile NaN setzen
    crosstab_perc = crosstab_abs.apply(lambda x: (x / row_sums[x.name] * 100) if row_sums[x.name] > 0 else np.nan, axis=1)

    # Optional: Länder nach Gesamtanzahl sortieren (oder nach einem anderen Kriterium)
    country_counts_sorted_index = df_clean['B_COUNTRY_ALPHA'].value_counts().index
    crosstab_perc_sorted = crosstab_perc.reindex(country_counts_sorted_index)
    # Entferne Zeilen, die nur NaN enthalten (Länder ohne gültige Altersdaten)
    crosstab_perc_sorted.dropna(how='all', inplace=True)


    # --- Visualisierung: Heatmap ---
    if not crosstab_perc_sorted.empty:
        plt.figure(figsize=(12, 10)) # Grösse anpassen je nach Anzahl Länder
        sns.heatmap(
            crosstab_perc_sorted,
            annot=True,       # Zeige die Prozentwerte in den Zellen
            fmt=".1f",        # Formatierung der Werte (eine Nachkommastelle)
            linewidths=.5,    # Linien zwischen Zellen
            cmap="viridis",   # Farbpalette (andere Optionen: "Blues", "YlGnBu", "coolwarm", etc.)
            cbar_kws={'label': 'Anteil der Befragten (%)'} # Beschriftung der Farbleiste
        )
        plt.title('Heatmap der relativen Altersgruppenverteilung pro Land', fontsize=16)
        plt.xlabel('Altersgruppe', fontsize=12)
        plt.ylabel('Land (ALPHA-Code)', fontsize=12)
        plt.xticks(rotation=45, ha='right') # X-Achsen-Labels lesbarer machen
        plt.yticks(rotation=0) # Y-Achsen-Labels horizontal
        plt.tight_layout()
        plt.show()
    else:
        print("Keine ausreichenden Daten zur Visualisierung der Altersgruppenverteilung pro Land vorhanden.")

    print("-" * 70)
```

#### t-Test, Q163 vs. Geschlecht

```python
# 2. Q163 vs. Geschlecht (Q260)
print("\n2. Q163 nach Geschlecht:")
if 'Geschlecht_Label' in df_clean.columns and 'Q163' in df_clean.columns:
    mean_q163_by_gender = df_clean.groupby('Geschlecht_Label')['Q163'].mean().round(2)
    print("Mittlere Q163-Bewertung pro Geschlecht:")
    # Stelle sicher, dass die Reihenfolge korrekt ist
    print(mean_q163_by_gender.loc[['Männlich', 'Weiblich']])

    plt.figure(figsize=(7, 5))
    sns.boxplot(x='Geschlecht_Label', y='Q163', data=df_clean, order=['Männlich', 'Weiblich'], palette='coolwarm')
    plt.title('Bewertung von Wissenschaft/Technik (Q163) nach Geschlecht')
    plt.xlabel('Geschlecht')
    plt.ylabel('Q163 Bewertung (1-10)')
    plt.ylim(0, 11)
    plt.tight_layout()
    plt.show()

    # T-Test (prüft, ob der Mittelwertunterschied statistisch signifikant ist)
    # Achtung: Setzt Normalverteilung und Varianzhomogenität voraus (visuell prüfen oder testen)
    try:
        from scipy.stats import ttest_ind
        group1 = df_clean[df_clean['Geschlecht_Label'] == 'Männlich']['Q163'].dropna()
        group2 = df_clean[df_clean['Geschlecht_Label'] == 'Weiblich']['Q163'].dropna()
        stat, p_value = ttest_ind(group1, group2, equal_var=False) # Welch's t-test (sicherer, wenn Varianzen ungleich)
        print(f"T-Test für Q163 zwischen Geschlechtern: Statistik={stat:.2f}, p-Wert={p_value:.3f}")
        if p_value < 0.05:
            print("Der Unterschied zwischen den Geschlechtern ist statistisch signifikant (p < 0.05).")
        else:
            print("Der Unterschied zwischen den Geschlechtern ist statistisch nicht signifikant (p >= 0.05).")
    except ImportError:
        print("Scipy nicht installiert, T-Test übersprungen.")
    except Exception as e:
        print(f"Fehler beim T-Test: {e}")

else:
    print("Benötigte Spalten ('Geschlecht_Label', 'Q163') nicht vollständig vorhanden.")
```

**Resultat**

Interpretation des ``T-Tests (Q163 nach Geschlecht)``  
Mittlere Q163-Bewertung pro Geschlecht: Zeigt den Durchschnittswert der Antworten auf Frage Q163 (auf der Skala 1-10) getrennt für Männer (7.30) und Frauen (7.21). Im Durchschnitt bewerten Männer die Auswirkungen von Wissenschaft und Technik also minimal höher als Frauen in deiner Stichprobe.
T-Test für Q163 zwischen Geschlechtern: Das ist ein statistischer Test, der prüft, ob dieser beobachtete Unterschied in den Mittelwerten (7.30 vs. 7.21) wahrscheinlich nur zufällig entstanden ist oder ob er auf einen tatsächlichen Unterschied in der Grundgesamtheit (also allen Menschen, die durch deine Stichprobe repräsentiert werden) hindeutet.
Statistik=2.73: Der T-Wert. Er gibt an, wie gross der Unterschied zwischen den Mittelwerten im Verhältnis zur Streuung innerhalb der Gruppen ist. Ein grösserer (absoluter) T-Wert deutet auf einen deutlicheren Unterschied hin.
p-Wert=0.006: Das ist die Wahrscheinlichkeit, einen solchen (oder noch grösseren) Unterschied zwischen den Gruppen zufällig zu beobachten, wenn es in Wirklichkeit gar keinen Unterschied zwischen den Geschlechtern gäbe (Nullhypothese).
Der Unterschied ... ist statistisch signifikant (p < 0.05): Weil der p-Wert (0.006) kleiner ist als das übliche Signifikanzniveau (0.05 oder 5%), gehen wir davon aus, dass der beobachtete Unterschied nicht nur Zufall ist. Wir schliessen daraus, dass es einen "echten" (statistisch signifikanten) Unterschied in der durchschnittlichen Bewertung zwischen Männern und Frauen in der Grundgesamtheit gibt.  

``Welche Hypothese wurde getestet?``  
Der T-Test prüft die Nullhypothese (H0) gegen eine Alternativhypothese (H1).
H0 (Nullhypothese): Es gibt keinen Unterschied im mittleren Q163-Wert zwischen Männern und Frauen in der Grundgesamtheit. (Der Mittelwert ist für beide Gruppen gleich).
H1 (Alternativhypothese): Es gibt einen Unterschied im mittleren Q163-Wert zwischen Männern und Frauen in der Grundgesamtheit. (Die Mittelwerte sind nicht gleich).
Fazit: Da p < 0.05, verwerfen wir die Nullhypothese (H0). Wir haben statistische Evidenz dafür gefunden, dass sich die mittleren Bewertungen von Männern und Frauen unterscheiden.

**Wichtig: Statistische Signifikanz bedeutet nicht unbedingt praktische Relevanz.** Der Unterschied ist zwar statistisch nachweisbar, aber mit 7.30 vs. 7.21 ist er in der Praxis sehr klein.

2. Interpretation der ``Pearson Korrelation`` (Alter Q262 vs. Q163)  
Korrelationskoeffizient: 0.034: Misst die Stärke und Richtung eines linearen Zusammenhangs zwischen dem Alter (Q262) und der Bewertung (Q163). Ein Wert nahe 0 bedeutet einen sehr schwachen linearen Zusammenhang. Der positive Wert deutet einen minimalen Trend an: "höheres Alter geht mit minimal höherer Bewertung einher".
p-Wert: 0.000: (oft < 0.001). Die Wahrscheinlichkeit, eine Korrelation von 0.034 (oder stärker) zufällig zu beobachten, wenn es in Wirklichkeit gar keinen linearen Zusammenhang gäbe.
**Die Korrelation ist statistisch signifikant (p < 0.05):** Weil der p-Wert sehr klein ist, schliessen wir, dass die beobachtete Korrelation (auch wenn sie winzig ist) wahrscheinlich nicht genau Null ist in der Grundgesamtheit. Es gibt statistische Evidenz für einen (extrem schwachen) linearen Zusammenhang.
Widerspruch "signifikant" aber "sehr schwach"? Ein zentraler Punkt in der Statistik: Statistische Signifikanz ≠ Stärke des Effekts (oder praktische Relevanz).
Statistische Signifikanz sagt uns nur, ob ein Effekt (Unterschied, Zusammenhang) wahrscheinlich "echt" ist (also nicht nur Zufall). Bei sehr grossen Stichproben (wie dieser mit über 21'000 Fällen) können auch extrem kleine, praktisch unbedeutende Effekte statistisch signifikant werden. Das liegt daran, dass der Test sehr viel "Power" hat, um selbst minimale Abweichungen von der Nullhypothese (kein Zusammenhang) zu entdecken.
Stärke des Effekts wird durch den Korrelationskoeffizienten selbst (0.034) beschrieben. Dieser Wert sagt uns, wie stark der (lineare) Zusammenhang ist. Und 0.034 ist eben sehr, sehr schwach.

**Fazit:** Das Alter hat zwar einen statistisch nachweisbaren linearen Einfluss auf die Bewertung Q163, aber dieser Einfluss ist so gering, dass er praktisch kaum eine Rolle spielt. Das Wissen über das Alter einer Person hilft uns kaum dabei, ihre Antwort auf Q163 vorherzusagen (zumindest nicht linear).

``Allg. zur Normalverteilung für diesen Datensatz``  
Die Normalverteilung wurde nicht explizit geprüft oder hergestellt.
T-Test: Ist relativ robust gegenüber Verletzungen der Normalverteilungsannahme, besonders bei grossen Stichproben wie dieser. Der Zentrale Grenzwertsatz sorgt dafür, dass die Verteilung der Mittelwerte sich einer Normalverteilung annähert, auch wenn die Rohdaten nicht normalverteilt sind. Daher ist **das Ergebnis des T-Tests hier wahrscheinlich trotzdem zuverlässig.**
**Pearson Korrelation: Setzt für die Signifikanzprüfung eigentlich auch Normalverteilung voraus. Aber auch hier gilt: Bei grossen Stichproben sind die Ergebnisse oft noch brauchbar.** Wichtiger ist aber: Pearson misst nur lineare Zusammenhänge. Wenn der Zusammenhang nicht-linear ist (z.B. U-förmig), wäre die Korrelation gering, obwohl ein Zusammenhang besteht. Das Streudiagramm hilft hier bei der Beurteilung.

#### Chi-Quadrat Test A: Land vs. Altersgruppe

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# --- Annahme: df_clean ist dein bereinigter DataFrame ---
# --- Stellt sicher, dass 'Altersgruppe_Label' existiert ---
if 'Altersgruppe_Label' not in df_clean.columns and 'X003R' in df_clean.columns:
    age_labels = {1.0: '16-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54', 5.0: '55-64', 6.0: '65+'}
    df_clean['Altersgruppe_Label'] = df_clean['X003R'].map(age_labels)
    print("Spalte 'Altersgruppe_Label' wurde erstellt.")
# --- Ende Label-Erstellung ---


print("\n--- Chi-Quadrat-Tests ---")

# Land vs. Altersgruppe
print("\nTest A: Land vs. Altersgruppe (alle Länder):\n")
if 'B_COUNTRY_ALPHA' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    # Finde Top 15 Länder
    countries = df_clean['B_COUNTRY_ALPHA'].value_counts().index.tolist()
    df_countries = df_clean[df_clean['B_COUNTRY_ALPHA'].isin(countries)].dropna(subset=['B_COUNTRY_ALPHA', 'Altersgruppe_Label'])

    # Erstelle die Kontingenztafel (Kreuztabelle)
    contingency_table_country_age = pd.crosstab(df_countries['B_COUNTRY_ALPHA'], df_countries['Altersgruppe_Label'])
    print("Kontingenztafel (Land vs. Altersgruppe):")
    print(contingency_table_country_age)

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2, p, dof, expected = chi2_contingency(contingency_table_country_age)
        print(f"\nChi-Quadrat Statistik: {chi2:.2f}")
        print(f"Freiheitsgrade (dof): {dof}")
        print(f"p-Wert: {p:.3g}") # .3g für wissenschaftliche Notation bei sehr kleinen p-Werten

        if p < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p < 0.05). Die Altersstrukturen unterscheiden sich.")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p >= 0.05).")
            
        # Warnung bei niedrigen erwarteten Häufigkeiten (kann Test ungültig machen)
        if (expected < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5. Das Ergebnis des Chi-Quadrat-Tests könnte unzuverlässig sein.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected < 5).sum()}")


    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('B_COUNTRY_ALPHA', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

**Resultat: Chi-Quadrat Land vs. Altersgruppe**  
Dieser Test prüft, ob die Altersverteilung (Altersgruppe_Label) in verschiedenen Ländern (B_COUNTRY_ALPHA) unterschiedlich ist.  

``Nullhypothese (H0):`` Es gibt keinen Zusammenhang zwischen dem Land und der Altersgruppe. Die prozentuale Verteilung der Altersgruppen ist in allen betrachteten Ländern gleich.  
``Alternativhypothese (H1):`` Es gibt einen Zusammenhang. Die Altersstruktur unterscheidet sich zwischen den Ländern.  
Wir führen den Test für die Top 15 Länder durch, um sicherzustellen, dass wir genügend Datenpunkte pro Zelle haben.
  
**Unterschiedliche Altersstrukturen nach Land:**  
Die Analyse der Teilnehmenden aus den Top-15-Ländern ergab einen signifikanten Zusammenhang zwischen dem Herkunftsland und der Altersgruppe (χ²(65) = 3354.05, p < 0.001). Dies bedeutet, dass sich die Altersverteilung der Befragten zwischen den untersuchten Ländern **deutlich unterscheidet.** Die Zusammensetzung der Stichprobe hinsichtlich junger, mittelalter und älterer Personen ist somit nicht über die Länder hinweg homogen.

#### Chi-Quadrat Test B: Q163 kategorisiert vs. Altersgruppe

```python
# Q163 (kategorisiert) vs. Altersgruppe
print("\nTest B: Q163 (kategorisiert) vs. Altersgruppe:\n")

if 'Q163' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    # Erstelle Q163 Kategorien
    bins = [0, 4, 7, 10] # Grenzen: 1-4, 5-7, 8-10
    labels = ['Niedrig (1-4)', 'Mittel (5-7)', 'Hoch (8-10)']
    df_clean['Q163_Kat'] = pd.cut(df_clean['Q163'], bins=bins, labels=labels, right=True, include_lowest=True)
    print(f"Neue Spalte 'Q163_Kat' erstellt. Verteilung:")
    print(df_clean['Q163_Kat'].value_counts(dropna=False))

    # Entferne Zeilen mit fehlenden Werten in den relevanten Spalten
    df_test_c = df_clean.dropna(subset=['Q163_Kat', 'Altersgruppe_Label'])

    # Erstelle die Kontingenztafel
    contingency_table_q163_age = pd.crosstab(df_test_c['Q163_Kat'], df_test_c['Altersgruppe_Label'])
    print("\nKontingenztafel (Q163 Kategorie vs. Altersgruppe):")
    print(contingency_table_q163_age)
    
    # Visualisierung als gestapeltes Balkendiagramm (Prozente)
    # Berechne Prozentsätze pro Altersgruppe
    cont_table_percent = contingency_table_q163_age.apply(lambda x: x*100/sum(x), axis=0)
    print("\nProzentuale Verteilung pro Altersgruppe:")
    print(cont_table_percent.round(1))
    
    cont_table_percent.T.plot(kind='bar', stacked=True, figsize=(12, 7), 
                              colormap='viridis', rot=0) # rot=0 für horizontale Labels
    plt.title('Verteilung der Q163-Bewertung (kategorisiert) nach Altersgruppe')
    plt.xlabel('Altersgruppe')
    plt.ylabel('Prozent (%)')
    plt.legend(title='Q163 Kategorie', bbox_to_anchor=(1.05, 1), loc='upper left') # Legende ausserhalb
    plt.tight_layout()
    plt.show()

    # Führe den Chi-Quadrat-Test B durch
    try:
        chi2_c, p_c, dof_c, expected_c = chi2_contingency(contingency_table_q163_age)
        print(f"\nChi-Quadrat Statistik: {chi2_c:.2f}")
        print(f"Freiheitsgrade (dof): {dof_c}")
        print(f"p-Wert: {p_c:.3g}")

        if p_c < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Q163-Kategorie und Altersgruppe (p < 0.05). Die Meinungsverteilung unterscheidet sich.")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Q163-Kategorie und Altersgruppe (p >= 0.05).")
        
        if (expected_c < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5. Das Ergebnis des Chi-Quadrat-Tests könnte unzuverlässig sein.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected_c < 5).sum()}")
             
    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")

else:
    print("\nBenötigte Spalten ('Q163', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

#### Chi-Quadrat Test: Q190 vs. Altersgruppe

```python
from scipy.stats import chi2_contingency

print("\n--- Chi-Quadrat-Test: Q190 (kategorisiert) vs. Altersgruppe ---")

if 'Q190_Kat' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    df_test_chi_q190_age = df_clean.dropna(subset=['Q190_Kat', 'Altersgruppe_Label'])

    contingency_q190_age = pd.crosstab(df_test_chi_q190_age['Q190_Kat'], df_test_chi_q190_age['Altersgruppe_Label'])
    print("Kontingenztafel (Q190 Kategorie vs. Altersgruppe):")
    print(contingency_q190_age)

    # Visualisierung als gestapeltes Balkendiagramm (Prozente)
    cont_q190_age_perc = contingency_q190_age.apply(lambda x: x*100/sum(x), axis=0)
    print("\nProzentuale Verteilung pro Altersgruppe:")
    print(cont_q190_age_perc.round(1))

    cont_q190_age_perc.T.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='coolwarm', rot=0)
    plt.title('Verteilung der Q190-Bewertung (kategorisiert) nach Altersgruppe')
    plt.xlabel('Altersgruppe')
    plt.ylabel('Prozent (%)')
    plt.legend(title='Q190 Kategorie', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2_qa, p_qa, dof_qa, expected_qa = chi2_contingency(contingency_q190_age)
        print(f"\nChi-Quadrat Statistik: {chi2_qa:.2f}")
        print(f"Freiheitsgrade (dof): {dof_qa}")
        print(f"p-Wert: {p_qa:.3g}")

        if p_qa < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p < 0.05).")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p >= 0.05).")

        if (expected_qa < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected_qa < 5).sum()}")

    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('Q190_Kat', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

**Resultat und Ergänzungen**

*   **Normalverteilung ist eine Eigenschaft einzelner Variablen:** Ich prüft nicht, ob der *gesamte Datensatz* normalverteilt ist, sondern ob *spezifische, metrische (Zahlen-)Variablen* einer Normalverteilung folgen (oder zumindest annähernd).  
``Kategoriale Variablen (wie Land, Geschlecht, Altersgruppe) können per Definition nicht normalverteilt sein.``  

*   **Zweck der Prüfung:** Die Annahme der Normalverteilung ist vor allem für bestimmte **inferenzstatistische Verfahren** wichtig (wie t-Tests, ANOVA, lineare Regression), da deren mathematische Grundlagen oft darauf aufbauen.  
*   **Deskriptive Analyse:** Für die *deskriptive* Analyse (die ich bisher hauptsächlich gemacht habe) ist es wichtiger, die **tatsächliche Verteilung** einer Variablen zu beschreiben (z.B. "linkssteil", "rechtssteil", "mehrgipflig", "gleichverteilt", "konzentriert auf bestimmte Werte"), anstatt nur zu sagen "ist/ist nicht normalverteilt". Die Histogramme und Boxplots haben genau das getan.  
*   **Skalen (1-10):** Bei Skalen wie 1-10 ist eine perfekte Normalverteilung eher unwahrscheinlich. Sie sind "ordinal" (haben eine Reihenfolge) aber nicht "intervallskaliert" im strengen Sinne (der Abstand zwischen 1 und 2 ist nicht zwingend gleich dem Abstand zwischen 9 und 10). Die Verteilungen sind oft schief oder haben Peaks an den Enden.
*   **Grosse Stichproben:** Bei sehr grossen Stichproben wie meiner (N > 21.000) greift der ``Zentrale Grenzwertsatz``. Dieser besagt, dass sich die *Verteilung der Mittelwerte* aus vielen Stichproben einer Normalverteilung annähert, selbst wenn die ursprüngliche Variable nicht normalverteilt ist. Das macht viele statistische Tests (wie den t-Test oder die lineare Regression) relativ robust gegenüber Verletzungen der Normalverteilungsannahme, *wenn die Stichprobe gross genug ist*. Die Signifikanztests sind also meist trotzdem gültig.
*   **Ist es "erledigt"?** Für die **deskriptive Phase:** Ja, ich haben die Verteilungen wichtiger Variablen beschrieben. Nun weiss ich, dass Q163 und Q190 nicht perfekt normalverteilt sind. Man *könnte* für jede Variable formale Tests auf Normalverteilung (z.B. Kolmogorov-Smirnov, Shapiro-Wilk) durchführen, aber bei N>21000 werden diese fast immer signifikant ausfallen (also "nicht normalverteilt" anzeigen), selbst bei leichten Abweichungen. Aussagekräftiger ist oft die visuelle Inspektion (Histogramm) und die Prüfung der Voraussetzungen für den *konkreten* Test, den man anwenden will.

### Schritt 2: Funktion zur Datenvorbereitung

```python
# --- Funktion zur PCA-Datenvorbereitung (Version für 'recode_map') ---

from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer # Für zukünftige Imputationsoption
import pandas as pd
import numpy as np

def prepare_pca_data(df, var_list, block_name, recode_map=None, nan_handling='dropna', min_cases_per_var=10):
    """
    Bereitet Daten für eine PCA vor: Variablenauswahl, NaN-Behandlung, optionales *einheitliches* Umkodieren, Standardisierung.

    Args:
        df (pd.DataFrame): Der Eingabe-DataFrame (z.B. df_clean).
        var_list (list): Liste der Variablennamen (Strings) für diesen PCA-Block.
        block_name (str): Name des Blocks für informative Ausgaben.
        recode_map (dict, optional): Einheitliches Dictionary für die Umkodierung {alt: neu} für alle Variablen im Block.
                                      Default ist None (keine Umkodierung).
        nan_handling (str, optional): Methode zur NaN-Behandlung ('dropna' oder 'impute'). Default ist 'dropna'.
        min_cases_per_var (int, optional): Mindestanzahl Fälle pro Variable für Warnung. Default ist 10.

    Returns:
        tuple: (pd.DataFrame, StandardScaler, pd.DataFrame) ->
               (Standardisierte Daten, Scaler-Objekt, Verarbeitete unskalierte Daten) oder (None, None, None) bei Fehler.
    """
    print(f"\n--- Starte Datenvorbereitung für PCA: {block_name} ---")

    # 1. Variablen auswählen und prüfen
    missing_vars = [var for var in var_list if var not in df.columns]
    if missing_vars:
        print(f"!! Fehler ({block_name}): Folgende Variablen wurden nicht gefunden: {missing_vars}")
        return None, None, None

    df_pca = df[var_list].copy()
    print(f"-> Variablen ausgewählt ({len(df_pca.columns)}): {', '.join(var_list)}")
    initial_rows = len(df_pca)
    print(f"   Ursprüngliche Anzahl Zeilen: {initial_rows}")

    # 2. Fehlende Werte behandeln
    print(f"\n-> Behandle fehlende Werte (Methode: {nan_handling})...")
    rows_with_nan = df_pca.isnull().any(axis=1).sum()
    print(f"   Zeilen mit mind. einem NaN vor Behandlung: {rows_with_nan} ({rows_with_nan/initial_rows:.1%})")

    if nan_handling == 'dropna':
        df_pca.dropna(inplace=True)
        rows_after = len(df_pca)
        print(f"   {initial_rows - rows_after} Zeilen wegen fehlender Werte entfernt.")
        print(f"   Verbleibende Zeilen für PCA: {rows_after}")
    # elif nan_handling == 'impute': # Option für später
    #     imputer = SimpleImputer(strategy='median')
    #     df_pca_imputed = imputer.fit_transform(df_pca)
    #     df_pca = pd.DataFrame(df_pca_imputed, index=df_pca.index, columns=df_pca.columns)
    #     rows_after = len(df_pca)
    #     print(f"   Fehlende Werte wurden mit der Strategie '{imputer.strategy}' imputiert.")
    #     print(f"   Anzahl Zeilen für PCA: {rows_after}")
    else:
        print(f"!! Fehler ({block_name}): Ungültige nan_handling Methode '{nan_handling}'")
        return None, None, None

    if rows_after == 0:
        print(f"!! Fehler ({block_name}): Keine Daten übrig nach NaN-Behandlung!")
        return None, None, None
    if rows_after < len(var_list) * min_cases_per_var:
        print(f"!! Warnung ({block_name}): Nur {rows_after} Fälle übrig ({rows_after/len(var_list):.1f} pro Variable). PCA könnte instabil sein.")

    # 3. Variablen umkodieren (falls recode_map gegeben)
    if recode_map is not None:
        print("\n-> Kodiere Variablen einheitlich um...")
        try:
            for var in var_list:
                 if var in df_pca.columns:
                     df_pca[var] = df_pca[var].replace(recode_map)
            print("   Einheitliche Umkodierung abgeschlossen.")
        except Exception as e:
            print(f"!! Fehler ({block_name}) beim Umkodieren: {e}")
            return None, None, None
    else:
        print("\n-> Keine Umkodierung angefordert.")

    # Sicherstellen, dass alle Spalten numerisch sind
    try:
        df_pca = df_pca.astype(float)
    except Exception as e:
        print(f"!! Fehler ({block_name}) bei Konvertierung zu numerischem Typ: {e}")
        return None, None, None

    df_pca_processed = df_pca.copy() # Zustand vor dem Skalieren speichern

    # 4. Daten standardisieren
    print("\n-> Standardisiere Daten (Mittelwert=0, StdAbw=1)...")
    scaler = StandardScaler()
    df_pca_scaled_array = scaler.fit_transform(df_pca)
    df_pca_scaled = pd.DataFrame(df_pca_scaled_array, index=df_pca.index, columns=df_pca.columns)
    print("   Daten erfolgreich standardisiert.")

    print(f"--- Datenvorbereitung für {block_name} abgeschlossen ---")
    return df_pca_scaled, scaler, df_pca_processed
```

### Schritt: 4 Hauptkomponentenanalyse (PCA) + Auswertung

```python
# --- Schritt 1: PCA durchführen und erklärte Varianz analysieren ---

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Sicherstellen, dass pandas importiert ist

print("========================================================")
print("=== Starte PCA und Varianzanalyse für alle Blöcke ===")
print("========================================================")

pca_results = {} # Dictionary zum Speichern der PCA-Objekte
explained_variance = {} # Dictionary zum Speichern der Varianz-Infos

# Stelle sicher, dass prepared_pca_data existiert und gefüllt ist
if 'prepared_pca_data' not in locals() or not prepared_pca_data:
    print("Fehler: 'prepared_pca_data' Dictionary nicht gefunden oder leer. Bitte führe die Vorbereitung zuerst aus.")
else:
    for block_name, data_dict in prepared_pca_data.items():
        print(f"\n--- Analyse für Block: {block_name} ---")

        scaled_data = data_dict['scaled_data']
        if scaled_data.empty:
            print("!! Leere Daten für diesen Block. Überspringe PCA.")
            continue

        # 1. PCA mit allen Komponenten durchführen
        # n_components=None bedeutet, behalte alle Komponenten
        pca = PCA(n_components=None)
        pca.fit(scaled_data)

        # Speichere das PCA-Objekt
        pca_results[block_name] = pca

        # 2. Erklärte Varianz analysieren
        exp_var_ratio = pca.explained_variance_ratio_
        cum_exp_var_ratio = np.cumsum(exp_var_ratio)
        explained_variance[block_name] = {
            'explained_variance_ratio': exp_var_ratio,
            'cumulative_explained_variance_ratio': cum_exp_var_ratio
        }

        print(f"  Anzahl möglicher Komponenten: {pca.n_components_}")
        print(f"  Erklärte Varianz pro Komponente: {np.round(exp_var_ratio, 3)}")
        print(f"  Kumulative erklärte Varianz:     {np.round(cum_exp_var_ratio, 3)}")

        # 3. Scree Plot erstellen
        plt.figure(figsize=(10, 5))

        # Ellenbogen-Plot (Erklärte Varianz pro Komponente)
        plt.subplot(1, 2, 1)
        num_components = len(exp_var_ratio)
        plt.plot(range(1, num_components + 1), exp_var_ratio, marker='o', linestyle='--')
        plt.title(f'Scree Plot (Ellenbogen)\n{block_name}')
        plt.xlabel('Anzahl Komponenten')
        plt.ylabel('Erklärte Varianz pro Komponente')
        plt.xticks(range(1, num_components + 1))
        plt.grid(True)

        # Kumulative erklärte Varianz
        plt.subplot(1, 2, 2)
        plt.plot(range(1, num_components + 1), cum_exp_var_ratio, marker='o', linestyle='-')
        # Linie für häufige Schwellenwerte (z.B. 80%, 90%, 95%)
        plt.axhline(y=0.95, color='r', linestyle=':', linewidth=1, label='95% Varianz')
        plt.axhline(y=0.90, color='g', linestyle=':', linewidth=1, label='90% Varianz')
        plt.axhline(y=0.80, color='b', linestyle=':', linewidth=1, label='80% Varianz')
        plt.title(f'Kumulative erklärte Varianz\n{block_name}')
        plt.xlabel('Anzahl Komponenten')
        plt.ylabel('Kumulative erklärte Varianz')
        plt.xticks(range(1, num_components + 1))
        plt.yticks(np.arange(0, 1.1, 0.1))
        plt.legend(loc='best')
        plt.grid(True)

        plt.tight_layout()
        plt.show()

print("\n========================================================")
print("=== PCA und Varianzanalyse abgeschlossen ===")
print("========================================================")
```

### Schritt 5: PCA mit gewählter Komponentenzahl + Drilldown

```python
# --- Schritt 2: PCA mit gewählter Komponentenzahl durchführen, Scores extrahieren & Ladungen vorbereiten ---

from sklearn.decomposition import PCA
import pandas as pd
import numpy as np

# 1. Definiere die Anzahl der Komponenten pro Block (basierend auf deiner Entscheidung)
num_components_to_keep = {
    "Block 1: Familie": 4,
    "Block 2: Erziehungswerte": 8,
    "Block 3: Soziale Distanz": 3,
    "Block 4: Grundbedürfnissen": 2,
    "Block 5: Science and Technology": 3
}

print("========================================================")
print("=== Starte finale PCA, Extraktion der Scores & Ladungsanalyse ===")
print("========================================================")

final_pca_models = {} # Zum Speichern der finalen PCA-Modelle (mit Ladungen)
principal_component_scores = {} # Zum Speichern der PC-Scores pro Block

# Stelle sicher, dass prepared_pca_data existiert
if 'prepared_pca_data' not in locals() or not prepared_pca_data:
    print("Fehler: 'prepared_pca_data' Dictionary nicht gefunden. Bitte führe die Vorbereitung zuerst aus.")
else:
    # Iteriere durch die Blöcke, für die Daten vorbereitet wurden
    for block_name in prepared_pca_data.keys():
        print(f"\n--- Finale PCA für Block: {block_name} ---")

        if block_name not in num_components_to_keep:
            print(f"!! Warnung: Anzahl Komponenten für '{block_name}' nicht definiert. Überspringe.")
            continue

        k = num_components_to_keep[block_name]
        scaled_data = prepared_pca_data[block_name]['scaled_data']
        original_vars = scaled_data.columns # Namen der Originalvariablen

        print(f"  Behalte k = {k} Komponenten.")

        # 2. PCA mit k Komponenten durchführen
        pca_final = PCA(n_components=k)
        pc_scores = pca_final.fit_transform(scaled_data)

        # Speichere das Modell
        final_pca_models[block_name] = pca_final

        # 3. PC-Scores als DataFrame speichern
        pc_col_names = [f'{block_name}_PC{i+1}' for i in range(k)]
        pc_scores_df = pd.DataFrame(data=pc_scores, columns=pc_col_names, index=scaled_data.index)
        principal_component_scores[block_name] = pc_scores_df
        print(f"  PC-Scores extrahiert und gespeichert (Shape: {pc_scores_df.shape}).")
        # print(pc_scores_df.head()) # Optional: Erste Zeilen der Scores anzeigen

        # 4. Ladungen analysieren (Vorbereitung - Ausgabe der Ladungen)
        print(f"\n  Ladungen (Loadings) für {block_name}:")
        # Ladungen sind in pca_final.components_ (Zeilen = PCs, Spalten = Originalvars)
        # Wir transponieren es für bessere Lesbarkeit (Zeilen=Originalvars, Spalten=PCs)
        loadings = pca_final.components_.T
        loadings_df = pd.DataFrame(loadings, index=original_vars, columns=[f'PC{i+1}' for i in range(k)])

        # Zeige Ladungen an (größte absolute Werte pro PC hervorheben)
        print("  (Positive Ladung: Variable korreliert positiv mit PC; Negative Ladung: negativ)")
        # Formatierung für bessere Lesbarkeit
        styled_loadings = loadings_df.style.format("{:.3f}").background_gradient(cmap='RdBu', axis=None) # Farbliche Hervorhebung
        display(styled_loadings) # display() funktioniert gut in Jupyter Notebooks

print("\n========================================================")
print("=== Finale PCA und Score-Extraktion abgeschlossen ===")
print("========================================================")

# Die principal_component_scores enthalten jetzt die neuen Variablen für jeden Block.
# Die final_pca_models enthalten die Ladungen zur Interpretation.
# Nächster Schritt: Interpretation der Ladungen & Zusammenführen der Scores.
```

#### Schritt 7: Imputation

22% der Daten zu verlieren ist mir zu viel.  
Da die PCs aber standardisiert sind (Mittelwert oft nahe 0), ist die Imputation mit dem Mittelwert oft eine akzeptable Lösung.

```python
# --- Schritt 4: NaN-Behandlung durch Imputation (Mittelwert) ---

from sklearn.impute import SimpleImputer
import pandas as pd

print("========================================================")
print("=== Starte NaN-Behandlung für Clustering/Regression ===")
print("========================================================")

# Stelle sicher, dass df_all_pca_scores existiert
if 'df_all_pca_scores' not in locals() or df_all_pca_scores.empty:
    print("Fehler: 'df_all_pca_scores' nicht gefunden oder leer.")
else:
    print(f"Ursprüngliche Shape vor Imputation: {df_all_pca_scores.shape}")
    nan_rows_before = df_all_pca_scores.isnull().any(axis=1).sum()
    print(f"Anzahl Zeilen mit NaN vor Imputation: {nan_rows_before}")

    # Initialisiere den Imputer (ersetzt NaN durch den Mittelwert der Spalte)
    imputer = SimpleImputer(strategy='mean')

    # Führe die Imputation durch
    # fit_transform gibt ein NumPy-Array zurück
    imputed_data_array = imputer.fit_transform(df_all_pca_scores)

    # Konvertiere zurück zu einem DataFrame mit den ursprünglichen Spalten und Index
    df_pca_scores_imputed = pd.DataFrame(
        imputed_data_array,
        columns=df_all_pca_scores.columns,
        index=df_all_pca_scores.index
    )

    print(f"\nShape nach Imputation: {df_pca_scores_imputed.shape}")
    nan_rows_after = df_pca_scores_imputed.isnull().any(axis=1).sum()
    print(f"Anzahl Zeilen mit NaN nach Imputation: {nan_rows_after}") # Sollte 0 sein

    print("\nErste 5 Zeilen des imputierten DataFrames:")
    print(df_pca_scores_imputed.head())

    # Optional: Prüfe, ob wirklich keine NaNs mehr da sind
    # print("\nPrüfung auf verbleibende NaNs pro Spalte:")
    # print(df_pca_scores_imputed.isnull().sum())

print("\n========================================================")
print("=== NaN-Behandlung abgeschlossen ===")
print("========================================================")

# df_pca_scores_imputed ist jetzt bereit für Clustering oder Regression.
```

## Cluster Analyse (k-means)

### Schritt 1: Optimale Clusteranzahl bestimmen

```python
# --- Schritt 1: Bestimmung der optimalen Clusteranzahl (Ellenbogen-Methode) ---

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd

print("========================================================")
print("=== Bestimme optimale Clusteranzahl (Ellenbogen-Methode) ===")
print("========================================================")

# Stelle sicher, dass die imputierten PCA-Scores existieren
if 'df_pca_scores_imputed' not in locals() or df_pca_scores_imputed.empty:
    print("Fehler: 'df_pca_scores_imputed' nicht gefunden oder leer.")
    print("Bitte führe Schritt 4 (Imputation) zuerst aus.")
else:
    # Datensatz für Clustering
    cluster_data = df_pca_scores_imputed

    # Wähle einen Bereich für K (Anzahl der Cluster), der getestet werden soll
    # Z.B. von 2 bis 15 Cluster. Bei sehr vielen Daten kann man auch höher gehen.
    k_range = range(2, 16)
    wcss = [] # Within-Cluster Sum of Squares (Inertia)

    print(f"Berechne K-Means für K von {min(k_range)} bis {max(k_range)}...")

    # Berechne K-Means für jeden Wert von K
    for k in k_range:
        kmeans = KMeans(n_clusters=k,
                        init='k-means++',   # Standard und meist gute Initialisierung
                        n_init='auto',      # Verhindert Warnung, passt Anzahl Initialisierungen an
                        random_state=42)    # Für reproduzierbare Ergebnisse
        kmeans.fit(cluster_data)
        wcss.append(kmeans.inertia_) # inertia_ ist der WCSS-Wert
        print(f"  K={k}, WCSS={kmeans.inertia_:.2f}")

    # --- Erstelle den Ellenbogen-Plot ---
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, wcss, marker='o', linestyle='-')
    plt.title('Ellenbogen-Methode zur Bestimmung der optimalen Clusteranzahl (K)')
    plt.xlabel('Anzahl der Cluster (K)')
    plt.ylabel('Within-Cluster Sum of Squares (WCSS / Inertia)')
    plt.xticks(list(k_range)) # Stelle sicher, dass alle K-Werte angezeigt werden
    plt.grid(True, linestyle='--')

    # Optional: Hervorheben des "Ellenbogens" - oft schwer automatisch zu finden
    # Manchmal wird die "KneeLocator"-Bibliothek verwendet, aber visuell ist oft besser.
    print("\nBitte untersuche den folgenden Plot:")
    print("Suche nach dem 'Ellenbogen': dem Punkt, an dem die Kurve abflacht")
    print("und das Hinzufügen eines weiteren Clusters die WCSS nicht mehr stark reduziert.")

    plt.show()

print("\n========================================================")
print("=== Ellenbogen-Analyse abgeschlossen ===")
print("========================================================")
# Nächster Schritt: Wähle K und führe das finale Clustering durch.
```

### Schritt 2: K-Means Clustering (K=5)

```python
# --- Schritt 2: Finales K-Means Clustering (K=5) ---

from sklearn.cluster import KMeans
import pandas as pd

print("========================================================")
print("=== Führe finales K-Means Clustering mit K=5 durch ===")
print("========================================================")

# Deine gewählte Anzahl Cluster
optimal_k = 5

# Stelle sicher, dass die imputierten PCA-Scores existieren
if 'df_pca_scores_imputed' not in locals() or df_pca_scores_imputed.empty:
    print("Fehler: 'df_pca_scores_imputed' nicht gefunden oder leer.")
    print("Bitte führe Schritt 4 (Imputation) zuerst aus.")
else:
    # Daten für das Clustering
    cluster_data = df_pca_scores_imputed

    # Initialisiere und führe K-Means mit K=5 durch
    kmeans_final = KMeans(n_clusters=optimal_k,
                          init='k-means++',
                          n_init='auto',
                          random_state=42) # Wichtig für Reproduzierbarkeit

    # Führe das Clustering durch und erhalte die Cluster-Labels
    cluster_labels = kmeans_final.fit_predict(cluster_data)

    # Füge die Cluster-Labels zum DataFrame mit den PCA-Scores hinzu
    # Erstelle eine Kopie, um das Original nicht zu verändern (optional, aber gute Praxis)
    df_clustered = df_pca_scores_imputed.copy()
    df_clustered['Cluster_K5'] = cluster_labels # Cluster-Labels beginnen bei 0

    print(f"K-Means Clustering mit K={optimal_k} abgeschlossen.")
    print("Cluster-Labels wurden zur Spalte 'Cluster_K5' hinzugefügt.")

    # Zeige die Größe der einzelnen Cluster an
    print("\nGröße der Cluster:")
    print(df_clustered['Cluster_K5'].value_counts().sort_index())

    # Optional: Zeige die ersten Zeilen mit Cluster-Labels
    print("\nErste 5 Zeilen des DataFrames mit Cluster-Labels:")
    print(df_clustered.head())

print("\n========================================================")
print("=== Clustering abgeschlossen ===")
print("========================================================")

# df_clustered enthält jetzt die PCA-Scores UND die Cluster-Zugehörigkeit.
# Nächster Schritt: Cluster-Analyse und -Interpretation.
```

```python
# --- Schritt 7: Cluster-Profile analysieren (Separate & Kombinierte Radar Charts) ---

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print("========================================================")
print("=== Analysiere Cluster-Profile (Separate & Kombinierte Radar Charts) ===")
print("========================================================")

# Stelle sicher, dass df_clustered existiert
if 'df_clustered' not in locals() or df_clustered.empty:
    print("Fehler: 'df_clustered' nicht gefunden oder leer.")
    print("Bitte führe Schritt 6 (Finales Clustering) zuerst aus.")
else:
    # 1. Berechne die Cluster-Zentren (wie zuvor)
    pca_cols = [col for col in df_clustered.columns if col.startswith('Block')]
    cluster_centroids = df_clustered.groupby('Cluster_K5')[pca_cols].mean()
    num_clusters = len(cluster_centroids)

    print("Cluster-Zentren (Durchschnittliche PCA-Scores pro Cluster):")
    print(cluster_centroids.round(2))

    # --- 2. Vorbereitung für Radar Charts ---
    labels = np.array(cluster_centroids.columns)
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    angles += angles[:1] # Schließe den Kreis

    # Finde Min/Max über alle Centroids für eine einheitliche Skala
    min_val = cluster_centroids.min().min()
    max_val = cluster_centroids.max().max()
    y_min = np.floor(min_val * 2) / 2 - 0.5 # Etwas Puffer
    y_max = np.ceil(max_val * 2) / 2 + 0.5 # Etwas Puffer
    y_ticks = np.round(np.linspace(y_min, y_max, 5), 1) # Runde Ticks für Anzeige

    # --- 3. Erstelle Figur mit Subplots (z.B. 2 Zeilen, 3 Spalten für K=5) ---
    # Passe nrows/ncols an, falls K anders ist
    nrows = 2
    ncols = 3
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols * 6, nrows * 6),
                             subplot_kw=dict(polar=True))
    fig.suptitle('Radar Charts der Cluster-Profile', fontsize=18, y=1.02)
    axes = axes.flatten() # Einfachere Indizierung der Achsen

    # Farben für die Cluster
    colors = plt.cm.tab10(np.linspace(0, 1, num_clusters))

    # --- 4. Separate Radar Charts für jeden Cluster ---
    for i, (index, row) in enumerate(cluster_centroids.iterrows()):
        ax = axes[i]
        data = row.values.flatten().tolist()
        data += data[:1] # Kreis schließen

        ax.plot(angles, data, color=colors[i], linewidth=2, linestyle='solid')
        ax.fill(angles, data, color=colors[i], alpha=0.25)

        ax.set_xticks(angles[:-1])
        ax.set_xticklabels(labels, fontsize=8) # Kleinere Schrift für Labels
        ax.set_yticks(y_ticks)
        ax.set_yticklabels([f"{tick:.1f}" for tick in y_ticks])
        ax.set_ylim(y_min, y_max)
        ax.set_title(f"Cluster {index}", size=14, y=1.1) # Titel für jeden Subplot

    # --- 5. Kombinierter Radar Chart im letzten Subplot ---
    ax_combined = axes[num_clusters] # Der nächste freie Subplot
    for i, (index, row) in enumerate(cluster_centroids.iterrows()):
        data = row.values.flatten().tolist()
        data += data[:1] # Kreis schließen
        ax_combined.plot(angles, data, color=colors[i], linewidth=1.5, linestyle='solid', label=f"Cluster {index}")
        ax_combined.fill(angles, data, color=colors[i], alpha=0.1) # Sehr transparent

    ax_combined.set_xticks(angles[:-1])
    ax_combined.set_xticklabels(labels, fontsize=8)
    ax_combined.set_yticks(y_ticks)
    ax_combined.set_yticklabels([f"{tick:.1f}" for tick in y_ticks])
    ax_combined.set_ylim(y_min, y_max)
    ax_combined.set_title("Kombinierte Profile", size=14, y=1.1)
    ax_combined.legend(loc='upper right', bbox_to_anchor=(1.4, 1.15), fontsize='small')

    # --- 6. Layout anpassen und Anzeigen ---
    # Ggf. leere Subplots ausblenden, falls nrows*ncols > num_clusters + 1
    for j in range(num_clusters + 1, nrows * ncols):
        fig.delaxes(axes[j])

    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Platz für Haupttitel lassen
    plt.show()

    # (Die textuelle Interpretation wie zuvor bleibt gültig)
    print("\nInterpretation der Radar Charts:")
    print("- Jede Achse repräsentiert eine Hauptkomponente (PC).")
    print("- Die ersten Plots zeigen das Profil jedes Clusters einzeln.")
    print("- Der letzte Plot zeigt alle Profile zum direkten Vergleich.")
    print("- Werte weiter außen bedeuten einen höheren Score auf dieser PC-Achse.")
    print("- Vergleiche die 'Formen': Wo haben die Cluster Spitzen? Wo sind sie eingezogen?")
    print("- Beziehe deine Interpretation der PCs mit ein, um die Cluster inhaltlich zu beschreiben.")

print("\n========================================================")
print("=== Cluster-Profil-Analyse (Radar Charts) abgeschlossen ===")
print("========================================================")
# Nächster Schritt: Detailliertere Cluster-Beschreibung & Vergleich mit Demografie.
```
