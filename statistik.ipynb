**2. QQ-Plots, t-Tests, Chi-Quadrat-Tests etc.**
Deine Verwirrung ist absolut verständlich, diese statistischen Tests können am Anfang überwältigend sein. Lass uns das kurz einordnen:

- **Was wir gerade machen:** Deskriptive Statistik. Wir beschreiben, wie die Daten aussehen (Häufigkeiten, Mittelwerte, Streuungen, Verteilungen durch Histogramme/Boxplots). Ziel ist es, ein Gefühl für die Daten zu bekommen.

- **Was diese Tests machen:** Inferenzstatistik. Sie werden verwendet, um Hypothesen zu testen oder Rückschlüsse von der Stichprobe auf die Grundgesamtheit zu ziehen.
  - **QQ-Plot:** Ein visuelles Werkzeug, um zu prüfen, ob Daten einer bestimmten theoretischen Verteilung folgen (oft Normalverteilung). Nützlich nachdem man sich die Verteilung z.B. im Histogramm angesehen hat, um die Abweichung von der Normalität genauer zu beurteilen. Könnten wir z.B. für Q262 (Alter) machen, um die Abweichung von der Normalverteilung deutlicher zu sehen.
  - **t-Test:** Vergleicht die Mittelwerte einer (ungefähr) normalverteilten Variable zwischen zwei Gruppen. Beispiel: "Ist der durchschnittliche Q163-Score bei Männern signifikant anders als bei Frauen?" (Voraussetzungen wie Normalverteilung innerhalb der Gruppen und Varianzhomogenität müssen geprüft werden, bei grossen Stichproben ist der Test aber robust gegen Verletzungen der Normalverteilungsannahme).
  - **Chi-Quadrat-Test (χ²):** Prüft, ob ein Zusammenhang zwischen zwei kategorialen Variablen besteht. Beispiel: "Gibt es einen Zusammenhang zwischen der Altersgruppe (X003R) und dem Land (B_COUNTRY_ALPHA)?" oder "Gibt es einen Zusammenhang zwischen Geschlecht (Q260) und einer kategorialen Antwort auf eine Frage?".
  
Wann verwenden? Diese Tests kommen meist nach der deskriptiven Analyse ins Spiel, wenn du spezifische Fragen oder Hypothesen hast, die du testen möchtest.
Beispiel: Nachdem wir sehen, dass Männer und Frauen leicht unterschiedliche Durchschnittswerte bei Q163 haben (deskriptiv), könnten wir einen t-Test verwenden, um zu prüfen, ob dieser Unterschied statistisch signifikant ist (inferenziell).
Beispiel: Nachdem wir sehen, dass die Verteilung der Altersgruppen je nach Land unterschiedlich aussieht (deskriptiv in Kreuztabellen), könnten wir einen Chi-Quadrat-Test verwenden, um zu prüfen, ob dieser Zusammenhang signifikant ist (inferenziell).

### Teil 2: Ausgewählte Q-Fragen

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np # Sicherstellen, dass numpy importiert ist

# --- Univariate Analyse - Teil 2: Ausgewählte Q-Fragen ---

print("--- Analyse ausgewählter Q-Fragen ---")

# Hilfsfunktion für die Analyse (jetzt mit optionalem ax Parameter)
# Gibt den (potenziell modifizierten) DataFrame zurück
def analyze_question(df, q_col, q_text, labels=None, scale_type='ordinal', ax=None):
    """
    Analysiert eine einzelne Frage (Spalte) im DataFrame.

    Args:
        df (pd.DataFrame): Der DataFrame.
        q_col (str): Der Spaltenname der Frage.
        q_text (str): Beschreibender Text für die Frage (für Titel).
        labels (dict, optional): Mapping von numerischen Werten zu Textlabels. Defaults to None.
        scale_type (str, optional): Typ der Skala ('ordinal', 'scale_1_10', 'binary'). Beeinflusst zusätzliche Plots. Defaults to 'ordinal'.
        ax (matplotlib.axes.Axes, optional): Achsenobjekt zum Plotten. Wenn None, wird eine neue Figur erstellt. Defaults to None.

    Returns:
        pd.DataFrame: Der potenziell modifizierte DataFrame (z.B. durch Typkonvertierung).
    """
    print(f"\n--- Analyse von {q_col}: {q_text} ---")
    if q_col not in df.columns:
        print(f"FEHLER: Spalte {q_col} nicht im DataFrame gefunden.")
        # Gebe das Original-df zurück, um Fehler im weiteren Verlauf zu vermeiden
        return df

    # Kopie erstellen, um SettingWithCopyWarning zu vermeiden und Original-df nicht zu ändern
    df_copy = df.copy()

    # Sicherstellen, dass die Spalte numerisch ist, falls nicht, versuchen zu konvertieren
    if not pd.api.types.is_numeric_dtype(df_copy[q_col]):
         try:
             # Versuche, in numerischen Typ zu konvertieren, erzwinge Fehler zu NaN
             df_copy[q_col] = pd.to_numeric(df_copy[q_col], errors='coerce')
             print(f"Hinweis: Spalte {q_col} wurde zu numerischem Typ konvertiert.")
         except Exception as e:
              print(f"FEHLER: Konnte Spalte {q_col} nicht in numerischen Typ konvertieren: {e}")
              # Gebe das Original-df zurück
              return df

    # Fehlende Werte zählen
    nan_count = df_copy[q_col].isnull().sum()
    total_count = len(df_copy[q_col])
    print(f"Anzahl fehlender Werte (NaN): {nan_count} (von {total_count} Einträgen)")

    # Value Counts (Häufigkeiten der Antworten)
    print("\nVerteilung der Antworten (Value Counts):")
    value_counts = df_copy[q_col].value_counts(dropna=False).sort_index() # dropna=False zeigt NaN count
    print(value_counts)

    # Deskriptive Statistik (nur sinnvoll für numerische Daten)
    print("\nDeskriptive Statistiken:")
    # Runden für bessere Lesbarkeit
    print(df_copy[q_col].describe().round(2))

    # --- Plotting ---
    # Entscheiden, ob eine neue Figur erstellt werden soll oder das übergebene ax verwendet wird
    create_new_figure = ax is None

    if create_new_figure:
        fig, current_ax = plt.subplots(figsize=(8, 5))
    else:
        current_ax = ax # Verwende das übergebene Axes-Objekt

    # Temporäre Label-Spalte nur erstellen, wenn Labels vorhanden sind
    temp_label_col = f'{q_col}_label_temp'
    plot_data = df_copy.dropna(subset=[q_col]) # Nur gültige Werte für Plot verwenden

    if labels:
        # Erstelle die temporäre Spalte nur im Plot-Daten-Frame
        plot_data = plot_data.copy() # Vermeide Warnung
        plot_data[temp_label_col] = plot_data[q_col].map(labels)

        # Sortiere die Labels basierend auf den ursprünglichen numerischen Werten
        valid_keys = sorted([k for k in labels.keys() if pd.notna(k)])
        ordered_labels = [labels[key] for key in valid_keys]

        sns.countplot(data=plot_data, x=temp_label_col, order=ordered_labels, palette='viridis', ax=current_ax)
        current_ax.set_xlabel("Antwort")
        # Ticks rotieren nur bei Textlabels
        current_ax.tick_params(axis='x', rotation=45, labelsize=9 if not create_new_figure else 10)

    else:
        # Direkt mit numerischen Werten plotten, wenn keine Labels vorhanden
        sns.countplot(data=plot_data, x=q_col, palette='viridis', ax=current_ax)
        current_ax.set_xlabel(f"Antwort ({q_col})")
        # Keine Rotation für numerische Ticks

    current_ax.set_ylabel("Anzahl")
    # Kürzeren Titel für Subplots, längeren für Einzelplots
    plot_title = f"{q_col}: {q_text}" if not create_new_figure else f"Verteilung der Antworten für {q_col}: {q_text}"
    current_ax.set_title(plot_title, fontsize=10 if not create_new_figure else 12)


    # Zusätzliche Plots (Histogramm, Boxplot) nur, wenn eine *neue* Figur erstellt wurde
    if create_new_figure:
        plt.tight_layout()
        plt.show() # Zeige den eben erstellten Plot

        # Zusätzliche Plots für Skalen (z.B. 1-10)
        if scale_type == 'scale_1_10':
            # Histogramm
            plt.figure(figsize=(10, 5))
            hist_data = df_copy[q_col].dropna()
            sns.histplot(hist_data, bins=10, kde=False)
            plt.title(f'Histogramm für {q_col}')
            plt.xlabel(f"Antwort ({q_col})")
            plt.ylabel('Häufigkeit')
            min_val, max_val = hist_data.min(), hist_data.max()
            if pd.notna(min_val) and pd.notna(max_val):
                 plt.xticks(range(int(min_val), int(max_val) + 1))
            plt.show()

            # Boxplot
            plt.figure(figsize=(8, 3))
            sns.boxplot(x=hist_data)
            plt.title(f'Boxplot für {q_col}')
            plt.xlabel(f"Antwort ({q_col})")
            plt.show()

    # Gib den DataFrame zurück, falls Typkonvertierungen stattfanden
    # Wichtig: Wenn Konvertierungen stattfanden, muss die Variable im Hauptcode neu zugewiesen werden
    return df_copy

# --- Fragen analysieren ---

# Stelle sicher, dass df_clean existiert und korrekt geladen/bereinigt wurde
# z.B. df_clean = pd.read_csv(...)
# ... Code zum Bereinigen ...

# WICHTIG: df_clean nach jedem Aufruf neu zuweisen, falls Typkonvertierungen nötig waren!
# Q1: Family Importance
q1_labels = {1.0: 'Sehr wichtig', 2.0: 'Eher wichtig', 3.0: 'Nicht sehr wichtig', 4.0: 'Gar nicht wichtig'}
df_clean = analyze_question(df_clean, 'Q1', 'Wichtigkeit der Familie', labels=q1_labels)

# Q6: Religion Importance
q6_labels = {1.0: 'Sehr wichtig', 2.0: 'Eher wichtig', 3.0: 'Nicht sehr wichtig', 4.0: 'Gar nicht wichtig'}
df_clean = analyze_question(df_clean, 'Q6', 'Wichtigkeit der Religion', labels=q6_labels)

# Q18-Q26: Unwanted Neighbors (Binär: 1=Genannt, 2=Nicht genannt) in einem Grid
neighbor_labels = {1.0: 'Genannt', 2.0: 'Nicht genannt'}
neighbor_questions = {
    'Q18': 'Drogenabhängige', 'Q19': 'Menschen anderer Rasse', 'Q20': 'Menschen mit AIDS',
    'Q21': 'Einwanderer/Gastarbeiter', 'Q22': 'Homosexuelle', 'Q23': 'Menschen anderer Religion',
    'Q24': 'Starke Trinker', 'Q25': 'Unverheiratete Paare', 'Q26': 'Menschen anderer Sprache'
}

# Erstelle eine 3x3 Figur für die Nachbarschaftsfragen
fig, axes = plt.subplots(3, 3, figsize=(15, 12), sharey=True) # sharey für gleichen Maßstab auf y-Achse
axes = axes.flatten() # Macht die 2D-Array zu 1D für einfaches Iterieren

print("\n--- Analyse von Q18-Q26: Unerwünschte Nachbarn (1=Genannt, 2=Nicht genannt) ---")
# Iteriere durch die Fragen und die Achsen
for i, (q_col, q_desc) in enumerate(neighbor_questions.items()):
    # Rufe analyze_question für die entsprechende Achse auf
    # WICHTIG: Hier df_clean NICHT neu zuweisen, da die Funktion nur zum Plotten dient
    # und wir nicht wollen, dass die Typkonvertierung 9x im Original-df versucht wird.
    # Die Textausgabe erfolgt trotzdem für jede Frage.
    analyze_question(df_clean, q_col, q_desc, labels=neighbor_labels, scale_type='binary', ax=axes[i])

# Layout anpassen und Gesamtfigur anzeigen
fig.suptitle("Ablehnung verschiedener Gruppen als Nachbarn (1=Genannt, 2=Nicht genannt)", fontsize=16, y=1.0) # y etwas höher
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # rect anpassen, um Platz für suptitle zu lassen
plt.show()


# Q46: Happiness
q46_labels = {1.0: 'Sehr glücklich', 2.0: 'Eher glücklich', 3.0: 'Nicht sehr glücklich', 4.0: 'Gar nicht glücklich'}
df_clean = analyze_question(df_clean, 'Q46', 'Allgemeines Glücksempfinden', labels=q46_labels)

# Q71: Confidence in Government
q71_labels = {1.0: 'Sehr grosses Vertrauen', 2.0: 'Ziemlich viel Vertrauen', 3.0: 'Nicht sehr viel Vertrauen', 4.0: 'Überhaupt kein Vertrauen'}
df_clean = analyze_question(df_clean, 'Q71', 'Vertrauen in die Regierung', labels=q71_labels)

# Q94: Church Membership
q94_labels = {0.0: 'Kein Mitglied', 1.0: 'Inaktives Mitglied', 2.0: 'Aktives Mitglied'}
df_clean = analyze_question(df_clean, 'Q94', 'Mitgliedschaft Kirche/Religionsgemeinschaft', labels=q94_labels)

# Q163: Impact of Science & Technology (Scale 1-10)
# Hier wird die Funktion wieder normal aufgerufen (ax=None), um Einzelplots zu erzeugen
df_clean = analyze_question(df_clean, 'Q163', 'Auswirkung Wissenschaft/Technologie (1=Viel schlechter, 10=Viel besser)', scale_type='scale_1_10')

# Q190: Justification Parents Beating Children (Scale 1-10)
df_clean = analyze_question(df_clean, 'Q190', 'Rechtfertigung Eltern schlagen Kinder (1=Nie, 10=Immer)', scale_type='scale_1_10')

print("\n--- Ende der Analyse ausgewählter Q-Fragen ---")
```

Welche Hypothese wurde getestet?
Der T-Test prüft die Nullhypothese (H0) gegen eine Alternativhypothese (H1).
H0 (Nullhypothese): Es gibt keinen Unterschied im mittleren Q163-Wert zwischen Männern und Frauen in der Grundgesamtheit. (Der Mittelwert ist für beide Gruppen gleich).

Fazit: Da p < 0.05, verwerfen wir die Nullhypothese (H0). Wir haben statistische Evidenz dafür gefunden, dass sich die mittleren Bewertungen von Männern und Frauen unterscheiden.
Wichtig: Statistische Signifikanz bedeutet nicht unbedingt praktische Relevanz. Der Unterschied ist zwar statistisch nachweisbar, aber mit 7.30 vs. 7.21 ist er in der Praxis sehr klein.
2. Interpretation der Pearson Korrelation (Alter Q262 vs. Q163)
Was wurde gemessen?
Korrelationskoeffizient: 0.034: Misst die Stärke und Richtung eines linearen Zusammenhangs zwischen dem Alter (Q262) und der Bewertung (Q163). Ein Wert nahe 0 bedeutet einen sehr schwachen linearen Zusammenhang. Der positive Wert deutet einen minimalen Trend an: "höheres Alter geht mit minimal höherer Bewertung einher".
p-Wert: 0.000: (Sehr klein, oft < 0.001). Die Wahrscheinlichkeit, eine Korrelation von 0.034 (oder stärker) zufällig zu beobachten, wenn es in Wirklichkeit gar keinen linearen Zusammenhang gäbe.
Die Korrelation ist statistisch signifikant (p < 0.05): Weil der p-Wert sehr klein ist, schliessen wir, dass die beobachtete Korrelation (auch wenn sie winzig ist) wahrscheinlich nicht genau Null ist in der Grundgesamtheit. Es gibt statistische Evidenz für einen (extrem schwachen) linearen Zusammenhang.
Widerspruch "signifikant" aber "sehr schwach"?
Das ist kein Widerspruch, sondern ein zentraler Punkt in der Statistik: Statistische Signifikanz ≠ Stärke des Effekts (oder praktische Relevanz).
Statistische Signifikanz sagt uns nur, ob ein Effekt (Unterschied, Zusammenhang) wahrscheinlich "echt" ist (also nicht nur Zufall). Bei sehr grossen Stichproben (wie deiner mit über 21'000 Fällen) können auch extrem kleine, praktisch unbedeutende Effekte statistisch signifikant werden. Das liegt daran, dass der Test sehr viel "Power" hat, um selbst minimale Abweichungen von der Nullhypothese (kein Zusammenhang) zu entdecken.
Stärke des Effekts wird durch den Korrelationskoeffizienten selbst (0.034) beschrieben. Dieser Wert sagt uns, wie stark der (lineare) Zusammenhang ist. Und 0.034 ist eben sehr, sehr schwach.
Fazit: Das Alter hat zwar einen statistisch nachweisbaren linearen Einfluss auf die Bewertung Q163, aber dieser Einfluss ist so gering, dass er praktisch kaum eine Rolle spielt. Das Wissen über das Alter einer Person hilft uns kaum dabei, ihre Antwort auf Q163 vorherzusagen (zumindest nicht linear).
3. Normalverteilung
Du hast Recht, wir haben die Normalverteilung nicht explizit geprüft oder hergestellt.
T-Test: Ist relativ robust gegenüber Verletzungen der Normalverteilungsannahme, besonders bei grossen Stichproben wie deiner. Der Zentrale Grenzwertsatz sorgt dafür, dass die Verteilung der Mittelwerte sich einer Normalverteilung annähert, auch wenn die Rohdaten nicht normalverteilt sind. Daher ist das Ergebnis des T-Tests hier wahrscheinlich trotzdem zuverlässig.
Pearson Korrelation: Setzt für die Signifikanzprüfung eigentlich auch Normalverteilung voraus. Aber auch hier gilt: Bei grossen Stichproben sind die Ergebnisse oft noch brauchbar. Wichtiger ist aber: Pearson misst nur lineare Zusammenhänge. Wenn der Zusammenhang nicht-linear ist (z.B. U-förmig), wäre die Korrelation gering, obwohl ein Zusammenhang besteht. Das Streudiagramm hilft hier bei der Beurteilung.
Wenn die Normalverteilung eine grosse Sorge wäre, könnten wir nicht-parametrische Tests verwenden (z.B. Mann-Whitney-U-Test statt T-Test, Spearman-Korrelation statt Pearson). Aber bei deiner Stichprobengrösse sind die aktuellen Ergebnisse wahrscheinlich okay.

1. Chi-Quadrat-Test: Land vs. Altersgruppe (Test b)
Dieser Test prüft, ob die Altersverteilung (Altersgruppe_Label) in verschiedenen Ländern (B_COUNTRY_ALPHA) unterschiedlich ist.
Nullhypothese (H0): Es gibt keinen Zusammenhang zwischen dem Land und der Altersgruppe. Die prozentuale Verteilung der Altersgruppen ist in allen betrachteten Ländern gleich.
Alternativhypothese (H1): Es gibt einen Zusammenhang. Die Altersstruktur unterscheidet sich zwischen den Ländern.
Wir führen den Test für die Top 15 Länder durch, um sicherzustellen, dass wir genügend Datenpunkte pro Zelle haben.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

# --- Annahme: df_clean ist dein bereinigter DataFrame ---
# --- Stelle sicher, dass 'Altersgruppe_Label' existiert ---
if 'Altersgruppe_Label' not in df_clean.columns and 'X003R' in df_clean.columns:
    age_labels = {1.0: '16-24', 2.0: '25-34', 3.0: '35-44', 4.0: '45-54', 5.0: '55-64', 6.0: '65+'}
    df_clean['Altersgruppe_Label'] = df_clean['X003R'].map(age_labels)
    print("Spalte 'Altersgruppe_Label' wurde erstellt.")
# --- Ende Label-Erstellung ---

print("\n--- Chi-Quadrat-Tests ---")

# Test b) Land vs. Altersgruppe
print("\nTest b) Land vs. Altersgruppe (Top 15 Länder):")
if 'B_COUNTRY_ALPHA' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    # Finde Top 15 Länder
    top_15_countries = df_clean['B_COUNTRY_ALPHA'].value_counts().nlargest(15).index.tolist()
    df_top15_countries = df_clean[df_clean['B_COUNTRY_ALPHA'].isin(top_15_countries)].dropna(subset=['B_COUNTRY_ALPHA', 'Altersgruppe_Label'])

    # Erstelle die Kontingenztafel (Kreuztabelle)
    contingency_table_country_age = pd.crosstab(df_top15_countries['B_COUNTRY_ALPHA'], df_top15_countries['Altersgruppe_Label'])
    print("Kontingenztafel (Land vs. Altersgruppe):")
    print(contingency_table_country_age)

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2, p, dof, expected = chi2_contingency(contingency_table_country_age)
        print(f"\nChi-Quadrat Statistik: {chi2:.2f}")
        print(f"Freiheitsgrade (dof): {dof}")
        print(f"p-Wert: {p:.3g}") # .3g für wissenschaftliche Notation bei sehr kleinen p-Werten

        if p < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p < 0.05). Die Altersstrukturen unterscheiden sich.")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Land und Altersgruppenverteilung (p >= 0.05).")
            
        # Warnung bei niedrigen erwarteten Häufigkeiten (kann Test ungültig machen)
        if (expected < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5. Das Ergebnis des Chi-Quadrat-Tests könnte unzuverlässig sein.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected < 5).sum()}")


    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('B_COUNTRY_ALPHA', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

Unterschiedliche Altersstrukturen nach Land (Test b): Die Analyse der Teilnehmenden aus den Top-15-Ländern ergab einen signifikanten Zusammenhang zwischen dem Herkunftsland und der Altersgruppe (χ²(65) = 3354.05, p < 0.001). Dies bedeutet, dass sich die Altersverteilung der Befragten zwischen den untersuchten Ländern deutlich unterscheidet. Die Zusammensetzung der Stichprobe hinsichtlich junger, mittelalter und älterer Personen ist somit nicht über die Länder hinweg homogen.

2. Chi-Quadrat-Test: Q163 (kategorisiert) vs. Altersgruppe (Test c)
Dieser Test prüft, ob die Verteilung der Meinungen zu Wissenschaft/Technik (in Kategorien eingeteilt) zwischen den Altersgruppen unterschiedlich ist.

Alternativhypothese (H1): Es gibt einen Zusammenhang. Die Meinungsverteilung unterscheidet sich zwischen den Altersgruppen.

```python
# Test c) Q163 (kategorisiert) vs. Altersgruppe
print("\nTest c) Q163 (kategorisiert) vs. Altersgruppe:")

if 'Q163' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    # Erstelle Q163 Kategorien
    bins = [0, 4, 7, 10] # Grenzen: 1-4, 5-7, 8-10
    labels = ['Niedrig (1-4)', 'Mittel (5-7)', 'Hoch (8-10)']
    df_clean['Q163_Kat'] = pd.cut(df_clean['Q163'], bins=bins, labels=labels, right=True, include_lowest=True)
    print(f"Neue Spalte 'Q163_Kat' erstellt. Verteilung:")
    print(df_clean['Q163_Kat'].value_counts(dropna=False))

    # Entferne Zeilen mit fehlenden Werten in den relevanten Spalten
    df_test_c = df_clean.dropna(subset=['Q163_Kat', 'Altersgruppe_Label'])

    # Erstelle die Kontingenztafel
    contingency_table_q163_age = pd.crosstab(df_test_c['Q163_Kat'], df_test_c['Altersgruppe_Label'])
    print("\nKontingenztafel (Q163 Kategorie vs. Altersgruppe):")
    print(contingency_table_q163_age)
    
    # Visualisierung als gestapeltes Balkendiagramm (Prozente)
    # Berechne Prozentsätze pro Altersgruppe
    cont_table_percent = contingency_table_q163_age.apply(lambda x: x*100/sum(x), axis=0)
    print("\nProzentuale Verteilung pro Altersgruppe:")
    print(cont_table_percent.round(1))
    
    cont_table_percent.T.plot(kind='bar', stacked=True, figsize=(12, 7), 
                              colormap='viridis', rot=0) # rot=0 für horizontale Labels
    plt.title('Verteilung der Q163-Bewertung (kategorisiert) nach Altersgruppe')
    plt.xlabel('Altersgruppe')
    plt.ylabel('Prozent (%)')
    plt.legend(title='Q163 Kategorie', bbox_to_anchor=(1.05, 1), loc='upper left') # Legende ausserhalb
    plt.tight_layout()
    plt.show()

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2_c, p_c, dof_c, expected_c = chi2_contingency(contingency_table_q163_age)
        print(f"\nChi-Quadrat Statistik: {chi2_c:.2f}")
        print(f"Freiheitsgrade (dof): {dof_c}")
        print(f"p-Wert: {p_c:.3g}")

        if p_c < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Q163-Kategorie und Altersgruppe (p < 0.05). Die Meinungsverteilung unterscheidet sich.")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Q163-Kategorie und Altersgruppe (p >= 0.05).")
        
        if (expected_c < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5. Das Ergebnis des Chi-Quadrat-Tests könnte unzuverlässig sein.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected_c < 5).sum()}")
             
    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")

else:
    print("\nBenötigte Spalten ('Q163', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

Okay, lass uns den Kruskal-Wallis-Test und das zugehörige Boxplot für Q190 (Rechtfertigung Eltern schlagen Kinder) Schritt für Schritt einordnen:

**1. Was macht der Kruskal-Wallis-Test?**

*   Dieser Test ist eine Alternative zur ANOVA, wenn die Daten nicht normalverteilt sind (was bei Skalen wie 1-10 oft der Fall ist).
*   Er prüft, ob sich die **zentrale Tendenz** (genauer gesagt: der **Median**) einer Variable (hier Q190) zwischen **mehreren Gruppen** (hier den 14 Ländern) statistisch signifikant unterscheidet.
*   **Nullhypothese (H0):** Der Median der Q190-Bewertung ist in allen 14 verglichenen Ländern gleich. Es gibt keinen Unterschied in der "typischen" Antwort zwischen den Ländern.
*   **Alternativhypothese (H1):** Mindestens ein Land hat einen anderen Median-Wert für Q190 als die anderen Länder.

**2. Interpretation deiner Ergebnisse:**

*   `Führe Kruskal-Wallis für 14 Länder mit >= 30 Antworten für Q190 durch:` Der Test wurde nur für die 14 Länder durchgeführt, für die du mindestens 30 gültige Antworten auf Frage Q190 hattest. Das stellt sicher, dass die Ergebnisse pro Land nicht auf zu wenigen Meinungen basieren.
*   `Kruskal-Wallis Statistik: 3310.40:` Das ist der berechnete Wert des Tests (oft als H-Statistik bezeichnet). Je grösser dieser Wert, desto stärker unterscheiden sich die Rangsummen (und damit tendenziell die Mediane) zwischen den Gruppen.
*   `p-Wert: 0:` Das ist der wichtigste Wert. Er gibt die Wahrscheinlichkeit an, solche (oder noch grössere) Unterschiede in den Medianen zwischen den Ländern zufällig zu finden, *wenn eigentlich die Nullhypothese wahr wäre* (also wenn in Wirklichkeit alle Länder den gleichen Median hätten). Ein p-Wert von 0 (oder extrem nahe bei 0) bedeutet, dass es äusserst unwahrscheinlich ist, diese Unterschiede nur durch Zufall zu erhalten.
*   `Ergebnis: Es gibt einen statistisch signifikanten Unterschied ... (p < 0.05):` Weil der p-Wert (0) kleiner als das übliche Signifikanzniveau von 0.05 ist, **verwerfen wir die Nullhypothese**. Das bedeutet: Der Test liefert starke statistische Evidenz dafür, dass sich die "typische" Antwort auf die Frage Q190 (Rechtfertigung von Kinderschlägen) zwischen den 14 untersuchten Ländern **tatsächlich unterscheidet**.

**3. Verbindung zum Boxplot (dem angehängten Bild):**

*   Das Boxplot ist die **visuelle Bestätigung** des Testergebnisses. Es zeigt *wie* sich die Verteilungen unterscheiden.
*   **Was du im Boxplot siehst:**
    *   **Jede Box** repräsentiert ein Land.
    *   Der **dicke Strich innerhalb jeder Box** ist der **Median** (der Wert, der die Antworten in eine obere und eine untere Hälfte teilt).
    *   Du siehst **deutliche Unterschiede bei den Medianen** zwischen den Ländern:
        *   Links im Bild (z.B. CAN, IRN, KGZ, NZL, UKR, TJK) liegen die Mediane sehr niedrig (oft bei 1 oder 2). Das bedeutet, in diesen Ländern ist die "typische" Antwort eher "niemals gerechtfertigt".
        *   Rechts im Bild (z.B. SGP, GTM, MNG, NIC, VNM, ZWE, MMR) liegen die Mediane deutlich höher (zwischen 3 und 5, bei MMR sogar bei 5). Hier ist die "typische" Antwort weniger ablehnend bzw. zeigt eine höhere Tendenz zur Rechtfertigung.
        *   ETH ist ein interessanter Fall mit einem hohen Median (ca. 5), aber auch einer grossen Streuung.
    *   Die **Höhe der Boxen** und die **Länge der "Whiskers"** (Linien ausserhalb der Box) zeigen die Streuung der Antworten *innerhalb* des Landes. Manche Länder haben sehr ähnliche Antworten (kleine Box), andere sehr unterschiedliche (grosse Box).
    *   Die **Punkte** sind Ausreisser (Antworten, die weit vom typischen Bereich entfernt liegen).

**Zusammenfassend:**

Der Kruskal-Wallis-Test sagt dir: "Ja, die Unterschiede zwischen den Ländern sind statistisch bedeutsam und nicht nur Zufall." Das Boxplot zeigt dir dann *visuell*, *welche* Länder niedrigere oder höhere Medianwerte haben und wie die Antworten innerhalb der Länder streuen.
```python
from scipy.stats import chi2_contingency

print("\n--- Chi-Quadrat-Test: Q190 (kategorisiert) vs. Altersgruppe ---")

if 'Q190_Kat' in df_clean.columns and 'Altersgruppe_Label' in df_clean.columns:
    df_test_chi_q190_age = df_clean.dropna(subset=['Q190_Kat', 'Altersgruppe_Label'])

    contingency_q190_age = pd.crosstab(df_test_chi_q190_age['Q190_Kat'], df_test_chi_q190_age['Altersgruppe_Label'])
    print("Kontingenztafel (Q190 Kategorie vs. Altersgruppe):")
    print(contingency_q190_age)

    # Visualisierung als gestapeltes Balkendiagramm (Prozente)
    cont_q190_age_perc = contingency_q190_age.apply(lambda x: x*100/sum(x), axis=0)
    print("\nProzentuale Verteilung pro Altersgruppe:")
    print(cont_q190_age_perc.round(1))

    cont_q190_age_perc.T.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='coolwarm', rot=0)
    plt.title('Verteilung der Q190-Bewertung (kategorisiert) nach Altersgruppe')
    plt.xlabel('Altersgruppe')
    plt.ylabel('Prozent (%)')
    plt.legend(title='Q190 Kategorie', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

    # Führe den Chi-Quadrat-Test durch
    try:
        chi2_qa, p_qa, dof_qa, expected_qa = chi2_contingency(contingency_q190_age)
        print(f"\nChi-Quadrat Statistik: {chi2_qa:.2f}")
        print(f"Freiheitsgrade (dof): {dof_qa}")
        print(f"p-Wert: {p_qa:.3g}")

        if p_qa < 0.05:
            print("-> Ergebnis: Es gibt einen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p < 0.05).")
        else:
            print("-> Ergebnis: Es gibt keinen statistisch signifikanten Zusammenhang zwischen Q190-Kategorie und Altersgruppe (p >= 0.05).")

        if (expected_qa < 5).sum() > 0:
             print("\nWarnung: Einige Zellen haben eine erwartete Häufigkeit < 5.")
             print(f"Anzahl Zellen mit erwarteter Häufigkeit < 5: {(expected_qa < 5).sum()}")

    except Exception as e:
        print(f"\nFehler beim Chi-Quadrat-Test: {e}")
else:
    print("\nBenötigte Spalten ('Q190_Kat', 'Altersgruppe_Label') nicht vollständig vorhanden.")
```

**1. Sache: Normalverteilung im Datensatz**

Du hast völlig Recht, die Frage nach der Normalverteilung ist klassisch und wichtig, aber oft auch verwirrend. Hier die Einordnung:

*   **Normalverteilung ist eine Eigenschaft einzelner Variablen:** Man prüft nicht, ob der *gesamte Datensatz* normalverteilt ist, sondern ob *spezifische, metrische (Zahlen-)Variablen* einer Normalverteilung folgen (oder zumindest annähernd). Kategoriale Variablen (wie Land, Geschlecht, Altersgruppe) können per Definition nicht normalverteilt sein.
*   **Zweck der Prüfung:** Die Annahme der Normalverteilung ist vor allem für bestimmte **inferenzstatistische Verfahren** wichtig (wie t-Tests, ANOVA, lineare Regression), da deren mathematische Grundlagen oft darauf aufbauen.
*   **Deskriptive Analyse:** Für die *deskriptive* Analyse (die wir bisher hauptsächlich gemacht haben) ist es wichtiger, die **tatsächliche Verteilung** einer Variablen zu beschreiben (z.B. "linkssteil", "rechtssteil", "mehrgipflig", "gleichverteilt", "konzentriert auf bestimmte Werte"), anstatt nur zu sagen "ist/ist nicht normalverteilt". Unsere Histogramme und Boxplots haben genau das getan.
*   **Deine Skalen (1-10):** Bei Skalen wie 1-10 (z.B. Q163, Q190) ist eine perfekte Normalverteilung eher unwahrscheinlich. Oft sind sie "ordinal" (haben eine Reihenfolge) aber nicht "intervallskaliert" im strengen Sinne (der Abstand zwischen 1 und 2 ist nicht zwingend gleich dem Abstand zwischen 9 und 10). Die Verteilungen sind oft schief oder haben Peaks an den Enden.
*   **Grosse Stichproben:** Bei sehr grossen Stichproben wie deiner (N > 21.000) greift der **Zentrale Grenzwertsatz**. Dieser besagt, dass sich die *Verteilung der Mittelwerte* aus vielen Stichproben einer Normalverteilung annähert, selbst wenn die ursprüngliche Variable nicht normalverteilt ist. Das macht viele statistische Tests (wie den t-Test oder die lineare Regression) relativ robust gegenüber Verletzungen der Normalverteilungsannahme, *wenn die Stichprobe gross genug ist*. Die Signifikanztests sind also meist trotzdem gültig.
*   **Ist es "erledigt"?** Für die **deskriptive Phase:** Ja, wir haben die Verteilungen wichtiger Variablen beschrieben. Wir wissen, dass z.B. Q163 und Q190 nicht perfekt normalverteilt sind. Für die **inferenzstatistische Phase** (wie die Tests, die wir gemacht haben): Wir haben uns (implizit) auf die Robustheit der Tests bei grossen Stichproben verlassen oder (wie beim Kruskal-Wallis-Test) eine Methode gewählt, die keine Normalverteilung voraussetzt. Man *könnte* für jede Variable formale Tests auf Normalverteilung (z.B. Kolmogorov-Smirnov, Shapiro-Wilk) durchführen, aber bei N>21000 werden diese fast immer signifikant ausfallen (also "nicht normalverteilt" anzeigen), selbst bei leichten Abweichungen. Aussagekräftiger ist oft die visuelle Inspektion (Histogramm) und die Prüfung der Voraussetzungen für den *konkreten* Test, den man anwenden will.

**Fazit zur Normalverteilung:** Du musst nicht den ganzen Datensatz auf Normalverteilung prüfen. Wir haben die Verteilungen wichtiger Variablen deskriptiv erfasst. Für die angewendeten Tests ist die Annahme entweder nicht zwingend erforderlich oder die Tests sind aufgrund der Stichprobengrösse robust. Du hast hier nichts Wichtiges übersehen.

```python
# Nur weitermachen, wenn df_scaled existiert und nicht leer ist
if 'df_scaled' in locals() and not df_scaled.empty:
    print("\n--- Bestimmung der optimalen Clusteranzahl (k) ---")
    wcss = []
    k_range = range(1, 11) # Teste k von 1 bis 10

    for k in k_range:
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, max_iter=300, random_state=42)
        kmeans.fit(df_scaled)
        wcss.append(kmeans.inertia_) # inertia_ ist die WCSS

    # Plotte die Ellbogenkurve
    plt.figure(figsize=(10, 6))
    plt.plot(k_range, wcss, marker='o', linestyle='--')
    plt.title('Ellbogenmethode zur Bestimmung von k')
    plt.xlabel('Anzahl der Cluster (k)')
    plt.ylabel('Within-Cluster Sum of Squares (WCSS)')
    plt.xticks(k_range)
    plt.grid(True)
    plt.show()

    print("Betrachte den Plot ('Ellbogenkurve'). Wo knickt die Kurve am deutlichsten ab?")
    print("Dieser 'Ellbogen' ist ein guter Kandidat für die optimale Anzahl der Cluster (k).")
    print("Wenn kein klarer Ellbogen sichtbar ist, sind die Daten möglicherweise nicht gut für K-Means geeignet oder es gibt keine deutliche Clusterstruktur.")

else:
     print("Keine skalierten Daten vorhanden, Bestimmung von k nicht möglich.")
```
